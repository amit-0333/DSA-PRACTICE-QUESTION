```python
## Brute Force
# TC: O(n^2)
# SC: O(n)  (visited array)

def mergeIntervals(intervals):
    n = len(intervals)
    intervals.sort()
    
    visited = [False] * n   
    
    res = []
    
    for i in range(n):
        if visited[i]:
            continue
        
        curr_start = intervals[i][0]
        curr_end = intervals[i][1]
        
        for j in range(i+1, n):
            if visited[j]:
                continue
            
            next_start = intervals[j][0]
            next_end = intervals[j][1]

            if next_start <= curr_end:
                curr_start = min(curr_start, next_start)
                curr_end = max(curr_end, next_end)
                visited[j] = True
        
        res.append([curr_start, curr_end])
        visited[i] = True
    
    return res


intervals = [[1,3],[2,6],[8,10],[15,18]]
print(mergeIntervals(intervals))
```

```python
## optimal solution
# Overall Time Complexity (TC): O(n log n)
# Space Complexity (SC): O(1) (excluding output list)

def mergeIntervals(intervals):
    intervals.sort()   # TC: O(n log n)
    
    res = []
    
    for curr in intervals:   # TC: O(n)
        curr_start = curr[0]
        curr_end = curr[1]

        # If no overlap OR result is empty → add interval
        if not res or curr_start > res[-1][1]:
            res.append([curr_start, curr_end])
        
        else:
            # Overlap → merge with last interval
            # Extend end to maximum
            res[-1][1] = max(res[-1][1], curr_end)
    
    return res


intervals = [[1,3],[2,6],[8,10],[15,18]]
print(mergeIntervals(intervals))
```
