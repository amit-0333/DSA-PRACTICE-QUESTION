```python
# Time Complexity (TC): O(n log n) → merge sort
# Space Complexity (SC): O(n) → temporary arrays

def mergeProcedure(arr, i, mid, j):

    count = 0

    nl = mid - i + 1
    nr = j - mid

    # temporary arrays
    ls = [0] * nl
    rs = [0] * nr

    # copy elements
    for m in range(nl):
        ls[m] = arr[i + m]
    for n in range(nr):
        rs[n] = arr[mid + 1 + n]

    # ✅ FIXED: correct counting logic (no modifying loop variable)
    right = mid + 1
    for left in range(i, mid + 1):   # FIXED variable name
        while right <= j and arr[left] > 2 * arr[right]:
            right += 1
        # count only valid elements
        count += (right - (mid + 1))   # FIXED formula

    # merge step
    p = 0
    q = 0
    k = i

    while p < nl and q < nr:
        if ls[p] <= rs[q]:
            arr[k] = ls[p]
            p += 1
        else:
            arr[k] = rs[q]
            q += 1
        k += 1

    # copy remaining elements
    while p < nl:
        arr[k] = ls[p]
        p += 1
        k += 1

    while q < nr:
        arr[k] = rs[q]
        q += 1
        k += 1

    return count


def mergeSort(arr, i, j):

    # base case
    if i >= j:
        return 0

    mid = i + (j - i) // 2

    # divide
    left_count = mergeSort(arr, i, mid)
    right_count = mergeSort(arr, mid + 1, j)

    # conquer (count + merge)
    curr_count = mergeProcedure(arr, i, mid, j)

    return left_count + right_count + curr_count


# Example
arr = [1, 3, 2, 3, 1]
print(mergeSort(arr, 0, len(arr) - 1))   # Output: 2
```
