```python
def majorityElement(arr):
    cand=None
    count=0
    for num in arr:
        if count==0:
            cand=num
        if num==cand:
            count+=1
        else:
            count-=1

    freq=0
    for num in arr:
        if num==cand:
            freq+=1
            
    if freq>len(arr)//2:
        return cand
    else:
        return -1


nums = [2,2,1,1,1,2,2]
print("majaority element is : " , majorityElement(nums))
```
