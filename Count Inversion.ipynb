```python
def mergeProcedure(arr, i, mid, j):
    # Time Complexity: O(n)
    # Space Complexity: O(n)

    curr_inv = 0

    nl = mid - i + 1
    nr = j - mid

    # temporary arrays
    ls = [0] * nl
    rs = [0] * nr

    # copy elements
    for m in range(nl):
        ls[m] = arr[i + m]
    for n in range(nr):
        rs[n] = arr[mid + 1 + n]

    p = 0
    q = 0
    k = i

    # merge + count inversions
    while p < nl and q < nr:
        if ls[p] <= rs[q]:
            arr[k] = ls[p]
            p += 1
        else:
            arr[k] = rs[q]
            curr_inv += (mid - i + 1) - p   # key step
            q += 1
        k += 1

    # copy remaining elements
    while p < nl:
        arr[k] = ls[p]
        p += 1
        k += 1

    while q < nr:
        arr[k] = rs[q]
        q += 1
        k += 1

    return curr_inv


def mergeSort(arr, i, j):
    # Time Complexity: O(n log n)
    # Space Complexity: O(n)

    if i >= j:
        return 0

    mid = i + (j - i) // 2

    left_inv = mergeSort(arr, i, mid)
    right_inv = mergeSort(arr, mid + 1, j)
    curr_inv = mergeProcedure(arr, i, mid, j)

    return left_inv + right_inv + curr_inv


# Example
arr = [2, 4, 1, 3, 5]
print(mergeSort(arr, 0, len(arr) - 1))  # Output: 3
```
