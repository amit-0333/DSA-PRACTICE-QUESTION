```python
def sort(arr):
    # Time Complexity: O(n)
    # Space Complexity: O(1)

    p0 = 0
    p2 = len(arr) - 1
    curr = 0

    while curr <= p2:
        if arr[curr] == 0:
            arr[curr], arr[p0] = arr[p0], arr[curr]
            p0 += 1
            curr += 1

        elif arr[curr] == 2:
            arr[curr], arr[p2] = arr[p2], arr[curr]
            p2 -= 1
           

        else:  # arr[curr] == 1
            curr += 1

    return arr


# Example
arr = [2,0,2,1,1,0]
print(sort(arr))
```
