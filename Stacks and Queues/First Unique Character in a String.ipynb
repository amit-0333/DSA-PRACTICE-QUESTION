```python
# TC: O(N)
# 2 passes over string

# SC: O(1)
# At most 26 characters (lowercase letters)

def firstUniqChar(string):
    
    h = {}
    
    # Count frequency
    for s in string:
        h[s] = h.get(s, 0) + 1
    
    # Find first unique character index
    for i in range(len(string)):
        if h[string[i]] == 1:
            return i
    
    return -1

test_cases = [
    "leetcode",
    "loveleetcode",
    "aabb",
    "abcabc",
    "z",
    ""
]

for s in test_cases:
    print(f"{s} -> {firstUniqChar(s)}")
```

```python
from collections import deque

# TC: O(N)
# Each element added and removed at most once

# SC: O(N)
# Queue + hashmap

def firstUniqChar(string: str) -> int:
    
    q = deque()
    h = {}
    
    for i, ch in enumerate(string):
        h[ch] = h.get(ch, 0) + 1
        q.append((ch, i))
        
        # Remove all non-unique from front
        while q and h[q[0][0]] > 1:
            q.popleft()
    
    # If queue has element → first unique
    return q[0][1] if q else -1

test_cases = [
    "leetcode",
    "loveleetcode",
    "aabb",
    "abcabc",
    "z",
    ""
]

for s in test_cases:
    print(f"{s} -> {firstUniqChar(s)}")
```
