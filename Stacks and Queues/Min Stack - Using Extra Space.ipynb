```python
# TC: O(1) for all operations
# push, pop, top, getMin are constant time

# SC: O(n)
# Extra stack to store minimums


class MinStack:

    # Approach:
    # Use two stacks:
    # 1. stack → stores all values
    # 2. minstack → stores minimum values
    #
    # push(val):
    #   - Push to stack
    #   - If minstack empty OR val <= current min → push to minstack
    #
    # pop():
    #   - Pop from stack
    #   - If popped element == minstack top → pop from minstack
    #
    # top():
    #   - Return top of stack
    #
    # getMin():
    #   - Return top of minstack (minimum element)

    def __init__(self):
        self.stack = []
        self.minstack = []

    def push(self, val: int) -> None:
        
        # Push to main stack
        self.stack.append(val)
        
        # Push to minstack if it's new minimum
        if not self.minstack or val <= self.minstack[-1]:
            self.minstack.append(val)

    def pop(self) -> None:
        
        # If stack empty
        if not self.stack:
            return
        
        # Pop element
        ele = self.stack.pop()
        
        # If it's the current minimum → remove from minstack
        if ele == self.minstack[-1]:
            self.minstack.pop()

    def top(self) -> int:
        
        # Return top of stack
        if not self.stack:
            return -1
        
        return self.stack[-1]

    def getMin(self) -> int:
        
        # Return current minimum
        if not self.minstack:
            return -1
        
        return self.minstack[-1]
```
