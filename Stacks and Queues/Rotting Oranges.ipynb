```python
from collections import deque

class Solution:

    # TC: O(N * M)
    # We visit each cell at most once during BFS
    
    # SC: O(N * M)
    # Queue + visited matrix in worst case

    def orangesRotting(self, grid):
        
        n = len(grid)
        m = len(grid[0])
        
        # Queue for BFS: stores ((row, col), time)
        q = deque()
        
        # Visited matrix to track processed cells
        v = [[False]*m for _ in range(n)]
        
        # Step 1: Push all initially rotten oranges into queue
        for i in range(n):
            for j in range(m):
                if grid[i][j] == 2:
                    q.append(((i, j), 0))   # multi-source BFS
                    v[i][j] = True
        
        ans = 0  # stores max time
        
        # Step 2: BFS traversal
        while q:
            (i, j), time = q.popleft()
            
            # Track maximum time
            ans = max(ans, time)
            
            # Explore 4 directions
            
            # UP
            if i-1 >= 0 and not v[i-1][j] and grid[i-1][j] == 1:
                q.append(((i-1, j), time+1))
                v[i-1][j] = True
            
            # DOWN
            if i+1 < n and not v[i+1][j] and grid[i+1][j] == 1:
                q.append(((i+1, j), time+1))
                v[i+1][j] = True
            
            # LEFT
            if j-1 >= 0 and not v[i][j-1] and grid[i][j-1] == 1:
                q.append(((i, j-1), time+1))
                v[i][j-1] = True
            
            # RIGHT
            if j+1 < m and not v[i][j+1] and grid[i][j+1] == 1:
                q.append(((i, j+1), time+1))
                v[i][j+1] = True
        
        # Step 3: Check if any fresh orange is left
        for i in range(n):
            for j in range(m):
                if grid[i][j] == 1 and not v[i][j]:
                    return -1
        
        return ans

# Driver Code
grid = [
    [2,1,1],
    [1,1,0],
    [0,1,1]
]

obj = Solution()
print(obj.orangesRotting(grid))  # Output: 4
```

```python
from collections import deque

class Solution:

    # TC: O(N * M)
    # Each cell is visited at most once
    
    # SC: O(N * M)
    # Queue + visited matrix

    def orangesRotting(self, grid):
        
        n = len(grid)
        m = len(grid[0])
        
        q = deque()
        v = [[False]*m for _ in range(n)]
        
        # Directions: UP, DOWN, LEFT, RIGHT
        directions = [(-1,0), (1,0), (0,-1), (0,1)]
        
        # Step 1: Push all rotten oranges
        for i in range(n):
            for j in range(m):
                if grid[i][j] == 2:
                    q.append(((i, j), 0))
                    v[i][j] = True
        
        ans = 0
        
        # Step 2: BFS
        while q:
            (i, j), time = q.popleft()
            ans = max(ans, time)
            
            # Iterate all 4 directions
            for di, dj in directions:
                ni = i + di
                nj = j + dj
                
                # Boundary + fresh + not visited
                if 0 <= ni < n and 0 <= nj < m:
                    if not v[ni][nj] and grid[ni][nj] == 1:
                        q.append(((ni, nj), time+1))
                        v[ni][nj] = True
        
        # Step 3: Check leftover fresh oranges
        for i in range(n):
            for j in range(m):
                if grid[i][j] == 1 and not v[i][j]:
                    return -1
        
        return ans

# Driver Code
grid = [
    [2,1,1],
    [1,1,0],
    [0,1,1]
]

obj = Solution()
print(obj.orangesRotting(grid))  # Output: 4
```
