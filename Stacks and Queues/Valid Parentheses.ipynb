```python
# TC: O(N)
# Traverse string once

# SC: O(N)
# Stack may store all opening brackets

def isValid(string: str) -> bool:
    
    stack = []
    
    # Mapping opening → closing
    bracketMapping = {
        '(': ')',
        '[': ']',
        '{': '}'
    }
    
    openParams = ('(', '[', '{')
    
    for s in string:
        
        # If opening bracket → push
        if s in openParams:
            stack.append(s)
        
        # If closing matches top
        elif stack and s == bracketMapping[stack[-1]]:
            stack.pop()
        
        # Invalid case
        else:
            return False
    
    # Valid if stack empty
    return len(stack) == 0

test_cases = ["()", "()[]{}", "(]", "([)]", "{[]}", "", "((()))"]

for s in test_cases:
    print(f"{s} -> {isValid(s)}")
```
