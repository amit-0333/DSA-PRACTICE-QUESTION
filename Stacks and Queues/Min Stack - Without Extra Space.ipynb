```python
class MinStack:

    # TC: O(1) for push(), pop(), top(), getMin()
    # SC: O(1) extra space
    # (We are NOT using an extra min stack. Only one stack + one variable)

    def __init__(self):
        # Main stack to store values (encoded or normal)
        self.stack = []
        
        # Variable to store current minimum element
        self.minele = None


    def push(self, val: int) -> None:
        """
        Approach:
        1. If stack is empty:
            - Push value normally
            - Set minele = val
        
        2. If new value is smaller than current min:
            - Push encoded value: (2*val - minele)
            - Update minele = val
        
        3. Else:
            - Push normally
        
        Why encoding?
        - Helps us store previous minimum implicitly
        - If a value < minele appears later → it's encoded
        """

        # Case 1: First element
        if not self.stack:
            self.stack.append(val)
            self.minele = val

        # Case 2: New minimum
        elif val < self.minele:
            # Encode the value before pushing
            encoded = 2 * val - self.minele
            self.stack.append(encoded)

            # Update current minimum
            self.minele = val

        # Case 3: Normal value
        else:
            self.stack.append(val)


    def pop(self) -> None:
        """
        Approach:
        1. If stack empty → return -1
        
        2. Pop top element:
            - If popped value < minele:
                → It was an encoded value
                → Decode previous minimum using:
                   previous_min = 2*minele - encoded_value
                → Update minele
        
            - Else:
                → Normal pop
        """

        if not self.stack:
            return -1

        top = self.stack.pop()

        # If encoded value
        if top < self.minele:
            # Recover previous minimum
            self.minele = 2 * self.minele - top


    def top(self) -> int:
        """
        Approach:
        1. If stack empty → return -1
        
        2. Check top element:
            - If top < minele:
                → It's encoded → actual value is minele
            
            - Else:
                → Return top normally
        """

        if not self.stack:
            return -1

        top = self.stack[-1]

        # Encoded case
        if top < self.minele:
            return self.minele

        return top


    def getMin(self) -> int:
        """
        Approach:
        - Simply return current minimum
        - If stack empty → return -1
        """

        if not self.stack:
            return -1

        return self.minele
```

```python
# Driver Code
obj = MinStack()

obj.push(5)
obj.push(3)
obj.push(7)
obj.push(2)

print("Current Min:", obj.getMin())   # 2
print("Top:", obj.top())              # 2

obj.pop()
print("After pop Min:", obj.getMin()) # 3

obj.pop()
print("Top:", obj.top())              # 3

obj.pop()
print("Min:", obj.getMin())           # 5
```
