```python
    # TC: O(N)
    # Single traversal of the arrays
    
    # SC: O(1)
    # Constant extra space
    
    def canCompleteCircuit(gas,cost):
        
        totalGas = 0
        totalCost = 0
        currGas = 0
        startGas = 0
        
        for i in range(len(gas)):
            totalGas += gas[i]
            totalCost += cost[i]
            
            currGas += gas[i] - cost[i]
            
            # If we can't reach i+1 from current start
            if currGas < 0:
                currGas = 0
                startGas = i + 1
        
        # Final feasibility check
        if totalGas < totalCost:
            return -1 
        
        return startGas

# Example input
gas = [1, 2, 3, 4, 5]
cost = [3, 4, 5, 1, 2]

# Function call
result = canCompleteCircuit(gas, cost)

# Output
print("Starting Index:", result)
```

```python

    # TC: O(N)
    # Single pass to check feasibility and find start index
    
    # SC: O(1)
    # No extra space used
    
    def canCompleteCircuit(gas,cost):
        
        totalGas = sum(gas)
        totalCost = sum(cost)

        # If total gas is less than total cost → impossible
        if totalGas < totalCost:
            return -1
        
        currGas = 0
        start = 0
        
        for i in range(len(gas)):
            currGas += gas[i] - cost[i]
            
            # If tank becomes negative → can't start from previous start
            if currGas < 0:
                currGas = 0
                start = i + 1   # move start to next index
        
        return start

# Example input
gas = [1, 2, 3, 4, 5]
cost = [3, 4, 5, 1, 2]

# Function call
result = canCompleteCircuit(gas, cost)

# Output
print("Starting Index:", result)
```
