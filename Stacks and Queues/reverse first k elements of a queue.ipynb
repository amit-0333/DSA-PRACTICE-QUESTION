```python
from collections import deque

# TC: O(N)
# SC: O(K)

def rev(q, k):
    
    n = len(q)
    
    # Edge cases
    if k > n or k <= 0:
        return q
    
    temp = []
    
    # Step 1: remove first k elements
    for _ in range(k):
        temp.append(q.popleft())
    
    # Step 2: push back in reverse order
    while temp:
        q.append(temp.pop())
    
    # Step 3: move remaining elements to back
    for _ in range(n - k):
        q.append(q.popleft())
    
    return q


q = deque([1,2,3,4,5])
k = 3

print("Original:", list(q))
print("Modified:", list(rev(q, k)))
```
