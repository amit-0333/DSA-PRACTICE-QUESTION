```python
def insertt(s, val):
    
    # TC: O(N)
    # In worst case, we may remove all elements to insert val
    
    # SC: O(N)
    # Recursive stack space
    
    """
    Approach:
    Insert 'val' into sorted stack 's' such that:
    - Top of stack always has largest element
    """

    # Base condition:
    # If stack is empty OR top element <= val → insert here
    if not s or s[-1] <= val:
        s.append(val)
        return
    
    # Otherwise remove top and recurse
    top = s.pop()
    insertt(s, val)
    
    # Put back removed element
    s.append(top)


def sortStack(s):
    
    # TC: O(N^2)
    # Each insertion takes O(N), and we do it N times
    
    # SC: O(N)
    # Due to recursion
    
    """
    Approach:
    1. Pop top element
    2. Recursively sort remaining stack
    3. Insert popped element at correct position using insertt()
    """

    # Base case
    if len(s) <= 1:
        return
    
    # Remove top
    top = s.pop()
    
    # Sort remaining stack
    sortStack(s)
    
    # Insert element in sorted order
    insertt(s, top)

# Driver Code
s = [3, 1, 4, 2]

print("Original Stack:", s)

sortStack(s)

print("Sorted Stack:", s)
```
