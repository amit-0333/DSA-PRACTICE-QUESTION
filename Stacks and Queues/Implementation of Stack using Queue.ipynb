```python
from collections import deque

class MyStack:

    # TC:
    # push: O(1)
    # pop: O(N)
    # top: O(N)
    # empty: O(1)

    # SC: O(N)

    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()
        
    def push(self, x: int) -> None:
        # Normal queue push
        self.q1.append(x)
    
    def pop(self) -> int:
        if not self.q1:
            return -1
        
        # Move all except last element
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        
        # Last element = stack top
        res = self.q1.popleft()
        
        # Swap queues
        self.q1, self.q2 = self.q2, self.q1
        
        return res

    def top(self) -> int:
        if not self.q1:
            return -1
        
        # Move all except last
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        
        # Peek last element
        res = self.q1[0]
        
        # Put it back
        self.q2.append(self.q1.popleft())
        
        # Swap back
        self.q1, self.q2 = self.q2, self.q1
        
        return res

    def empty(self) -> bool:
        return len(self.q1) == 0


# Driver code for Costly Pop Approach

obj = MyStack()

obj.push(10)
obj.push(20)
obj.push(30)

print("Top:", obj.top())     # Expected: 30
print("Pop:", obj.pop())     # Expected: 30
print("Top:", obj.top())     # Expected: 20
print("Empty:", obj.empty()) # Expected: False

print("Pop:", obj.pop())     # Expected: 20
print("Pop:", obj.pop())     # Expected: 10
print("Empty:", obj.empty()) # Expected: True
```

```python
from collections import deque

class MyStack:

    # TC:
    # push: O(N)  -> we move all elements from q1 to q2
    # pop: O(1)   -> direct popleft
    # top: O(1)   -> direct access
    # empty: O(1)
    
    # SC: O(N) -> storing all elements in queues

    def __init__(self):
        # q1 will always store elements in stack order
        # q2 is helper queue used during push
        self.q1 = deque()
        self.q2 = deque()
        
    def push(self, x: int) -> None:
        # Step 1: Insert new element into empty helper queue q2
        self.q2.append(x)
        
        # Step 2: Move ALL elements from q1 → q2
        # This ensures new element comes FIRST (stack top)
        while self.q1:
            self.q2.append(self.q1.popleft())
        
        # Step 3: Swap names of queues
        # Now q1 has updated order, q2 becomes empty
        self.q1, self.q2 = self.q2, self.q1
    
    def pop(self) -> int:
        # If stack is empty
        if not self.q1:
            return -1
        
        # Front of q1 is always the stack top
        return self.q1.popleft()
    
    def top(self) -> int:
        # If stack is empty
        if not self.q1:
            return -1
        
        # First element is top of stack
        return self.q1[0]
    
    def empty(self) -> bool:
        # Stack is empty if q1 has no elements
        return len(self.q1) == 0



# Driver code for Costly Push Approach

obj = MyStack()

obj.push(1)
obj.push(2)
obj.push(3)

print("Top:", obj.top())     # Expected: 3
print("Pop:", obj.pop())     # Expected: 3
print("Top:", obj.top())     # Expected: 2
print("Empty:", obj.empty()) # Expected: False

print("Pop:", obj.pop())     # Expected: 2
print("Pop:", obj.pop())     # Expected: 1
print("Empty:", obj.empty()) # Expected: True
```
