```python
def stockspan(arr):
    
    # TC: O(N)
    # Each element is pushed and popped at most once
    
    # SC: O(N)
    # Stack + result array
    
    """
    Approach:
    1. Find index of Nearest Greater to Left (NGL)
    2. If no greater element → store -1
    3. Span = i - NGL_index
    """

    stack = []   # stores (value, index)
    res = []
    n = len(arr)
    
    # Step 1: Find NGL indices
    for i in range(n):
        
        # Pop smaller/equal elements
        while stack and stack[-1][0] <= arr[i]:
            stack.pop()
        
        # If stack empty → no greater on left
        if not stack:
            res.append(-1)
        else:
            res.append(stack[-1][1])
        
        # Push current element
        stack.append((arr[i], i))
    
    # Step 2: Convert to span
    for i in range(n):
        res[i] = i - res[i]
    
    return res


# Input prices
prices = [100, 80, 60, 70, 60, 75, 85]
print(stockspan(prices))
```
