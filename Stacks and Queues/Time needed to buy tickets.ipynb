```python
from collections import deque

# TC: O(sum of tickets)
# Each ticket is processed once

# SC: O(n)
# Queue stores all people

def timeRequiredToBuy(tickets, k):
    
    # Initialize queue
    q = deque()
    
    # Add all people (tickets, index)
    for i in range(len(tickets)):
        q.append((tickets[i], i))
    
    time = 0  # total time
    
    # Process queue
    while q:
        
        # Take front person
        tickets_needed, person_idx = q.popleft()
        
        # One second passes
        time += 1
        
        # Person buys 1 ticket
        tickets_needed -= 1
        
        # If kth person finishes, return time
        if person_idx == k and tickets_needed == 0:
            return time
        
        # If still tickets left, push back
        if tickets_needed > 0:
            q.append((tickets_needed, person_idx))


#  DRIVER CODE 
tickets = [2, 3, 2]
k = 2

result = timeRequiredToBuy(tickets, k)
print(result)  # Expected Output: 6
```

```python
# TC: O(n)
# Single pass through the array

# SC: O(1)
# No extra space used

def timeRequiredToBuy(tickets, k):
    
    time = 0  # total time
    
    # Traverse all people
    for i in range(len(tickets)):
        
        # If person is before or equal to k
        if i <= k:
            # They contribute up to tickets[k]
            time += min(tickets[i], tickets[k])
        
        else:
            # They contribute up to tickets[k] - 1
            time += min(tickets[i], tickets[k] - 1)
    
    return time


#  DRIVER CODE 
tickets = [2, 3, 2]
k = 2

print(timeRequiredToBuy(tickets, k))  # Output: 6
```
