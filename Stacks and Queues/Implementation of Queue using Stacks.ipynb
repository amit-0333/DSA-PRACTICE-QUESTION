```python
from collections import deque

class MyQueue:

    # TC (Amortized):
    # push: O(1)
    # pop: O(1)
    # peek: O(1)
    # (Each element moves at most once from s1 → s2)

    # SC: O(N)

    def __init__(self):
        # s1 = input stack (push here)
        # s2 = output stack (pop/peek from here)
        self.s1 = deque()
        self.s2 = deque()

    def push(self, x):
        # Always push into s1
        self.s1.append(x)

    def pop(self):
        # If s2 is empty → transfer all elements from s1
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        
        # If still empty → queue is empty
        if not self.s2:
            return -1

        # Pop from s2 (front of queue)
        return self.s2.pop() 

    def peek(self):
        # Same logic as pop, but don't remove element
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        
        if not self.s2:
            return -1

        # Return front element
        return self.s2[-1] 

    def empty(self):
        # Queue is empty only if both stacks are empty
        return not self.s1 and not self.s2


# Driver code

obj = MyQueue()

obj.push(1)
obj.push(2)
obj.push(3)

print("Peek:", obj.peek())   # Expected: 1
print("Pop:", obj.pop())     # Expected: 1
print("Peek:", obj.peek())   # Expected: 2
print("Empty:", obj.empty()) # Expected: False

print("Pop:", obj.pop())     # Expected: 2
print("Pop:", obj.pop())     # Expected: 3
print("Empty:", obj.empty()) # Expected: True
```

```python
from collections import deque

class MyQueue:

    # TC:
    # push: O(1)
    # pop: O(N)
    # peek: O(N)
    # empty: O(1)

    # SC: O(N)

    def __init__(self):
        self.s1 = deque()
        self.s2 = deque()

    def push(self, x: int) -> None:
        self.s1.append(x)
        
    def pop(self) -> int:
        if not self.s1:
            return -1
        
        # Move all elements
        while self.s1:
            self.s2.append(self.s1.pop())
        
        res = self.s2.pop()
        
        # Move back
        while self.s2:
            self.s1.append(self.s2.pop())
        
        return res
        
    def peek(self) -> int:
        if not self.s1:
            return -1
        
        while self.s1:
            self.s2.append(self.s1.pop())
        
        res = self.s2[-1]
        
        # Move back
        while self.s2:
            self.s1.append(self.s2.pop())
        
        return res
        
    def empty(self) -> bool:
        return len(self.s1) == 0

# Driver code

obj = MyQueue()

obj.push(1)
obj.push(2)
obj.push(3)

print("Peek:", obj.peek())   # Expected: 1
print("Pop:", obj.pop())     # Expected: 1
print("Peek:", obj.peek())   # Expected: 2
print("Empty:", obj.empty()) # Expected: False

print("Pop:", obj.pop())     # Expected: 2
print("Pop:", obj.pop())     # Expected: 3
print("Empty:", obj.empty()) # Expected: True
```
