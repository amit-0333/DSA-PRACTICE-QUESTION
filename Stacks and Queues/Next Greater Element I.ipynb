```python
class Solution:

    def ngr(self, nums):
    
        # TC: O(N)
        # Each element is pushed and popped at most once
        
        # SC: O(N)
        # Stack + result array
        
        n = len(nums)
        res = []
        stack = []
        
        for i in range(n-1, -1, -1):
            
            if len(stack) == 0:
                res.append(-1)
            
            elif len(stack) > 0 and stack[-1] > nums[i]:
                res.append(stack[-1])
            
            elif len(stack) > 0 and stack[-1] <= nums[i]:
                
                while len(stack) > 0 and stack[-1] <= nums[i]:
                    stack.pop()
                
                if len(stack) == 0:
                    res.append(-1)
                else:
                    res.append(stack[-1])
            
            stack.append(nums[i])
        
        res.reverse()
        return res
        return res
    def nextGreaterElement(self, nums1, nums2):
    
        # TC: O(N + M)
        # ngr(nums2) -> O(N)
        # mapping -> O(N)
        # lookup for nums1 -> O(M)
        
        # SC: O(N)
        # map + result
        
        ngei = self.ngr(nums2)
        
        mp = {}
        
        # Map each element of nums2 to its NGE
        for i in range(len(nums2)):
            mp[nums2[i]] = ngei[i]
        
        res = []
        
        # Get result for nums1 from map
        for x in nums1:
            res.append(mp[x])
        
        return res

# Driver code

obj = Solution()

nums1 = [4, 1, 2]
nums2 = [1, 3, 4, 2]

result = obj.nextGreaterElement(nums1, nums2)

print("Result:", result)   # Expected: [-1, 3, -1]
```

```python
class Solution:

    def nextGreaterElement(self, nums1, nums2):
    
    # TC: O(N + M)
    # Each element processed once in nums2
    # Lookup for nums1 is O(M)
    
    # SC: O(N)
    # Stack + map
    
        stack = []
        mp = {}
        
        # Process nums2
        for num in nums2:
            
            # If current number is greater than stack top
            while len(stack) > 0 and num > stack[-1]:
                mp[stack[-1]] = num
                stack.pop()
            
            # Push current element
            stack.append(num)
        
        # Remaining elements → no greater element
        while len(stack) > 0:
            mp[stack[-1]] = -1
            stack.pop()
        
        # Build result for nums1
        res = []
        for x in nums1:
            res.append(mp[x])
        
        return res


# Driver code

obj = Solution()

nums1 = [4, 1, 2]
nums2 = [1, 3, 4, 2]

result = obj.nextGreaterElement(nums1, nums2)

print("Result:", result)   # Expected: [-1, 3, -1]
```
