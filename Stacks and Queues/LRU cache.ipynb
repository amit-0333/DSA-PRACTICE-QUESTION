```python
# Approach:
# We implement LRU Cache using a list.
#
# The list stores (key, value) pairs.
# Order of list represents usage:
# - Front  → Least Recently Used (LRU)
# - End    → Most Recently Used (MRU)
#
# Operations:
#
# 1. get(key):
#    - Search key in list
#    - If found:
#        → Remove it from current position
#        → Append it to end (mark as most recently used)
#        → Return value
#    - If not found:
#        → Return -1
#
# 2. put(key, value):
#    - If key already exists:
#        → Remove old entry
#        → Append updated (key, value) at end
#    - If key does not exist:
#        → Append (key, value)
#    - If size exceeds capacity:
#        → Remove front element (LRU)
#
# Note:
# This is NOT optimal (uses list → O(n) operations)
# Optimal solution uses HashMap + Doubly Linked List (O(1))

# TC:
# get  -> O(n)
# put  -> O(n)

# SC: O(capacity)


class LRUCache:

    def __init__(self, capacity: int):
        # Store maximum capacity of cache
        self.capacity = capacity
        
        # List to store (key, value) pairs
        self.cache = []

    def get(self, key: int) -> int:
        
        # Traverse cache to find the key
        for i in range(len(self.cache)):
            
            # Extract key and value
            stored_key, stored_val = self.cache[i]
            
            # If key found
            if stored_key == key:
                
                # Remove from current position
                self.cache.pop(i)
                
                # Append to end (mark as most recently used)
                self.cache.append((stored_key, stored_val))
                
                # Return value
                return stored_val
        
        # Key not found
        return -1

    def put(self, key: int, value: int) -> None:
        
        # Traverse to check if key already exists
        for i in range(len(self.cache)):
            
            stored_key, stored_val = self.cache[i]
            
            # If key exists
            if stored_key == key:
                
                # Remove old entry
                self.cache.pop(i)
                
                # Insert updated value at end (MRU)
                self.cache.append((key, value))
                
                return
        
        # If key is new → insert at end
        self.cache.append((key, value))
        
        # If capacity exceeded → remove LRU (front)
        if len(self.cache) > self.capacity:
            self.cache.pop(0)

#  DRIVER CODE
# Create cache with capacity 2
obj = LRUCache(2)

# Put operations
obj.put(1, 1)   # cache = [(1,1)]
obj.put(2, 2)   # cache = [(1,1), (2,2)]

# Get operation
print(obj.get(1))  # Output: 1
# cache becomes [(2,2), (1,1)]

# Insert new key → evicts LRU (key 2)
obj.put(3, 3)   # cache = [(1,1), (3,3)]

print(obj.get(2))  # Output: -1 (not found)

# Insert new key → evicts LRU (key 1)
obj.put(4, 4)   # cache = [(3,3), (4,4)]

print(obj.get(1))  # Output: -1
print(obj.get(3))  # Output: 3
print(obj.get(4))  # Output: 4
```

```python
from collections import OrderedDict

# TC: O(1) for get and put
# All operations (lookup, insert, delete, move) are constant time

# SC: O(capacity)
# Stores at most 'capacity' elements


class LRUCache:

    # Approach:
    # We use OrderedDict which maintains insertion order.
    #
    # Order meaning:
    # - Front  → Least Recently Used (LRU)
    # - End    → Most Recently Used (MRU)
    #
    # get(key):
    #   - If key not present → return -1
    #   - If present:
    #       → Move key to end (mark as recently used)
    #       → Return its value
    #
    # put(key, value):
    #   - If key exists:
    #       → Move it to end (update recent usage)
    #   - Insert/update the key with value
    #   - If size exceeds capacity:
    #       → Remove first element (LRU)

    def __init__(self, capacity: int):
        # Store capacity
        self.capacity = capacity
        
        # Initialize OrderedDict
        self.cache = OrderedDict()

    def get(self, key: int) -> int:
        
        # If key not found
        if key not in self.cache:
            return -1
        
        # Move key to end (MRU)
        self.cache.move_to_end(key)
        
        # Return value
        return self.cache[key]

    def put(self, key: int, value: int) -> None:
        
        # If key exists → move to end
        if key in self.cache:
            self.cache.move_to_end(key)

        # Insert or update value
        self.cache[key] = value
        
        # If capacity exceeded → remove LRU
        if len(self.cache) > self.capacity:
            self.cache.popitem(last=False)

#  DRIVER CODE
# Create cache with capacity 2
obj = LRUCache(2)

# Put operations
obj.put(1, 1)   # cache = [(1,1)]
obj.put(2, 2)   # cache = [(1,1), (2,2)]

# Get operation
print(obj.get(1))  # Output: 1
# cache becomes [(2,2), (1,1)]

# Insert new key → evicts LRU (key 2)
obj.put(3, 3)   # cache = [(1,1), (3,3)]

print(obj.get(2))  # Output: -1 (not found)

# Insert new key → evicts LRU (key 1)
obj.put(4, 4)   # cache = [(3,3), (4,4)]

print(obj.get(1))  # Output: -1
print(obj.get(3))  # Output: 3
print(obj.get(4))  # Output: 4
```
