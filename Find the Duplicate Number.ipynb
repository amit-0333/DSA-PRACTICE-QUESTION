```python
## Optimal solution
def findDuplicate(nums):
    # Time Complexity: O(n)
    # Reason: Both phases (cycle detection + finding entry) together take linear time

    # Space Complexity: O(1)
    # Reason: Only pointers are used, no extra data structures

    # Phase 1: Detect cycle using Floyd’s Tortoise & Hare
    slow = 0
    fast = 0

    while True:
        slow = nums[slow]          # move 1 step
        fast = nums[nums[fast]]    # move 2 steps

        if slow == fast:
            break   # cycle detected

    # Phase 2: Find entrance of cycle (duplicate number)
    slow2 = 0   # start from index 0 (IMPORTANT)

    while True:
        slow = nums[slow]
        slow2 = nums[slow2]

        if slow == slow2:
            return slow   # duplicate element

nums = [1,3,4,2,2]
print(findDuplicate(nums))
```

```python
def findRepeated(nums):
    # Time Complexity: O(n)
    # Reason: Single loop over array, set operations are O(1) on average

    # Space Complexity: O(n)
    # Reason: Set may store up to n elements in worst case
    
    seen = set()
    n = len(nums)
    repeated = -1

    for i in range(n):
        # check value instead of index
        if nums[i] in seen:
            repeated = nums[i]   # found repeated element
            break               # stop at first repetition
        else:
            seen.add(nums[i])   # store element

    return repeated

nums = [1,3,4,2,2]
print(findRepeated(nums))
```

```python
def findRepeated(nums):
    # Time Complexity: O(n log n)
    # Reason: Sorting takes O(n log n), loop takes O(n)

    # Space Complexity: O(1)
    # Reason: No extra space used (in-place sorting)

    nums.sort()  # sort the array so duplicates come together
    rep = -1

    # iterate till second last element to avoid index error
    for i in range(len(nums) - 1):
        # check adjacent elements
        if nums[i] == nums[i + 1]:
            rep = nums[i]  # duplicate found
            break

    return rep

nums = [1,3,4,2,2]
print(findRepeated(nums))
```
