```python
def merge2sortedarray(arr1, m, arr2, n):
    # Time Complexity: O(m + n)
    # Space Complexity: O(1)

    i = m - 1   # last element of arr1 (valid part)
    j = n - 1   # last element of arr2
    k = m + n - 1  # last index of arr1 (total size)

    # Merge from back
    while i >= 0 and j >= 0:
        if arr1[i] > arr2[j]:
            arr1[k] = arr1[i]
            i -= 1
        else:
            arr1[k] = arr2[j]
            j -= 1
        k -= 1

    # Copy remaining elements of arr2
    while j >= 0:
        arr1[k] = arr2[j]
        j -= 1
        k -= 1


# Example
arr1 = [1,2,3,0,0,0]
arr2 = [2,5,6]
m = 3
n = 3

merge2sortedarray(arr1, m, arr2, n)
print(arr1)
```
