```python
def maxProfit(arr):
    # Time Complexity: O(n)
    # Space Complexity: O(1)

    minPrice = float('inf')
    maxProfit = 0

    for i in range(len(arr)):
        # Update minimum price (buy point)
        if arr[i] < minPrice:
            minPrice = arr[i]

        # Calculate profit and update max
        elif arr[i] - minPrice > maxProfit:
            maxProfit = arr[i] - minPrice

    return maxProfit

prices = [7,1,5,3,6,4]
print(maxProfit(prices))
```
