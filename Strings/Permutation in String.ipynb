```python
def checkInclusion(s1, s2):
    # Time Complexity: O(n)
    # - n = len(s2)
    # - Sliding window over s2
    # - Each comparison is O(26) ≈ constant

    # Space Complexity: O(1)
    # - Two arrays of size 26

    # handle edge case (removes runtime error)
    if len(s1) > len(s2):
        return False

    # 🔹 Step 1: frequency arrays
    count1 = [0] * 26
    count2 = [0] * 26

    # 🔹 Step 2: count s1
    for ch in s1:
        count1[ord(ch) - ord('a')] += 1

    # 🔹 Step 3: first window
    for i in range(len(s1)):
        count2[ord(s2[i]) - ord('a')] += 1

    # 🔹 Step 4: check first window
    if count1 == count2:
        return True

    # 🔹 Step 5: sliding window
    for i in range(len(s1), len(s2)):
        count2[ord(s2[i]) - ord('a')] += 1
        count2[ord(s2[i - len(s1)]) - ord('a')] -= 1

        if count1 == count2:
            return True

    # 🔹 Step 6: no match
    return False



s1 = "ab"
s2 = "eidbaooo"
print(checkInclusion(s1, s2))   # True
```
