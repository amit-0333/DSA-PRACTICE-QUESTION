```python
# Time Complexity: O(n * m) + O(m^2) (due to string concatenation)
# Space Complexity: O(m) (result string)

def longestCommonPrefix(strs):
    # Edge case: if list is empty
    if not strs:
        return ''

    res = ''  # stores the common prefix

    # Iterate over characters of the first string
    for i in range(len(strs[0])):
        
        # Compare ith character with all strings
        for s in strs:
            # If index goes out of bounds OR characters mismatch
            if i >= len(s) or s[i] != strs[0][i]:
                return res  # return prefix found so far
        
        # If all strings have same character at index i
        res += strs[0][i]

    return res

strs = ["flower","flow","flight"]
print(longestCommonPrefix(strs))
```

```python
 # Time Complexity: O(n * m)
# Space Complexity: O(1) (excluding output), O(m) including output
def longestCommonPrefix(strs):
    # Edge case: if list is empty
    if not strs:
        return ''

    # Iterate over characters of the first string
    for i in range(len(strs[0])):

        # Compare ith character with all strings
        for s in strs:
            # If index goes out of bounds OR characters mismatch
            if i >= len(s) or s[i] != strs[0][i]:
                return strs[0][:i]  # return prefix till previous index

    # If loop completes, entire first string is common prefix
    return strs[0]

strs = ["flower","flow","flight"]
print(longestCommonPrefix(strs))
```
