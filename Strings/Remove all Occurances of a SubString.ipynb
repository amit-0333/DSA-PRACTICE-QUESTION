```python
def removeOccurrences(s, part):
    # Time Complexity: O(n * m)
    # - n = len(s), m = len(part)
    # - each replace scans string

    # Space Complexity: O(n)
    # - new string created each time

    while part in s:
        s = s.replace(part, '', 1)  

    return s

s = "daabcbaabcbc"
part = "abc"
print(removeOccurrences(s,part))
```

```python
def removeOccurrences(s, part):
    # Time Complexity: O(n*m) (due to join)
    # - Each character processed once (amortized)

    # Space Complexity: O(n)
    # - Stack stores characters

    stack = []
    m = len(part)

    for ch in s:
        stack.append(ch)

        # check last m characters
        if len(stack) >= m:
            if ''.join(stack[-m:]) == part:
                for _ in range(m):
                    stack.pop()

    return ''.join(stack)   

s = "daabcbaabcbc"
part = "abc"
print(removeOccurrences(s,part))
```

```python
def removeOccurrences(s, part):
    # Time Complexity: O(n)
    # Space Complexity: O(n)

    stack = []
    m = len(part)

    for ch in s:
        stack.append(ch)

        if len(stack) >= m:
            match = True

            # compare without join
            for i in range(m):
                if stack[len(stack) - m + i] != part[i]:
                    match = False
                    break

            if match:
                del stack[-m:]

    return ''.join(stack)
s = "daabcbaabcbc"
part = "abc"
print(removeOccurrences(s,part))
```
