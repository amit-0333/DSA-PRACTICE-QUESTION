```python
from collections import defaultdict

def groupAnagrams(strs):
    # Time Complexity: O(n * k)
    # n = number of strings, k = max length of a string
    # - For each string, we build a 26-length frequency array → O(k)
    
    # Space Complexity: O(n * k)
    # - Storing all strings in hashmap
    # - Keys are tuples of size 26

    res = defaultdict(list)  # key → frequency tuple, value → list of anagrams

    for s in strs:  
        count = [0] * 26     # frequency array for each string

        for ch in s.lower():         # count characters → O(k)
            count[ord(ch) - ord('a')] += 1

        res[tuple(count)].append(s)  # tuple used as hashable key

    return list(res.values())  # grouped anagrams

strs = ["eat","tea","tan","ate","nat","bat"]
print(groupAnagrams(strs))
```
