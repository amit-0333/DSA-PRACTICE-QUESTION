```python
def isAnagram(s, t):
    # Time Complexity: O(n)
    # - Single pass for s and t

    # Space Complexity: O(1)
    # - Fixed array of size 26

    if len(s) != len(t):
        return False

    freq = [0] * 26  # for 'a' to 'z'

    for ch in s:
        freq[ord(ch) - ord('a')] += 1

    for ch in t:
        idx = ord(ch) - ord('a')
        freq[idx] -= 1
        if freq[idx] < 0:
            return False

    return True

s = "anagram"
t = "nagaram"
print(isAnagram(s,t))
```
