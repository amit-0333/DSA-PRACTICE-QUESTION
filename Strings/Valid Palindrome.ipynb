```python
def isPalindrome(s):
    # Time Complexity: O(n)
    # - First loop takes O(n)
    # - Second loop takes O(n/2) ≈ O(n)
    
    # Space Complexity: O(n)
    # - Extra string `new` stores filtered characters

    new = ''
    
    # Build cleaned string (only alphanumeric, lowercase)
    for ch in s:
        if ch.isalnum():
            new += ch.lower()

    n = len(new)

    # Check palindrome
    for i in range(n // 2):
        if new[i] != new[n - i - 1]:
            return False

    return True

s = "A man, a plan, a canal: Panama"
print(isPalindrome(s))
```

```python
def isPalindrome(s):
    # Time Complexity: O(n)
    # - Each character is visited at most once by i and j

    # Space Complexity: O(1)
    # - No extra space used (in-place comparison)

    i = 0
    j = len(s) - 1

    while i < j:
        # Skip non-alphanumeric characters
        if not s[i].isalnum():
            i += 1
            continue

        if not s[j].isalnum():
            j -= 1
            continue

        # Compare after converting to lowercase
        if s[i].lower() != s[j].lower():
            return False

        i += 1
        j -= 1

    return True

s = "A man, a plan, a canal: Panama"
print(isPalindrome(s))
```
