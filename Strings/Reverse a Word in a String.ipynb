```python
def reverseWords(s):
    # Time Complexity: O(n)
    # - Each character is visited at most once

    # Space Complexity: O(n)
    # - 'res' stores the resulting words

    res = []                 # store words in reverse order
    i = len(s) - 1           # start from end of string

    while i >= 0:
        #  Step 1: skip spaces
        # move left until we find a non-space character
        while i >= 0 and s[i] == ' ':
            i -= 1

        # if we reached beginning, stop
        if i < 0:
            break

        #  Step 2: mark end of word
        j = i

        # Step 3: find start of word
        # move left until space or start
        while i >= 0 and s[i] != ' ':
            i -= 1

        #  Step 4: extract full word
        # word is from index (i+1) to j
        res.append(s[i+1:j+1])

        # now i is at space before word → next loop continues

    #  Step 5: join words with single space
    return " ".join(res)

s = "the sky is blue"
print(reverseWords(s))
```

```python
def reverseWords(s):
    # Time Complexity: O(n)
    # Space Complexity: O(n)

    return " ".join(s.split()[::-1])

s = "the sky is blue"
print(reverseWords(s))
```
