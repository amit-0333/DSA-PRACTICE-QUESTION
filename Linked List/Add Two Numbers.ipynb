```python
class Solution:
    
    # TC: O(max(N, M))
    # We traverse both lists once
    
    # SC: O(max(N, M))
    # New list is created for result
    
    def addTwoNumbers(self, l1, l2):
        
        # Create dummy node to build result list
        l3 = ListNode(0)
        head = l3   # store starting point
        
        carry = 0
        
        # Traverse both lists together
        while l1 and l2:
            
            # Add values with carry
            value = l1.val + l2.val + carry
            
            # Update carry
            carry = value // 10
            
            # Create new node with digit
            l3.next = ListNode(value % 10)
            
            # Move pointers
            l3 = l3.next
            l1 = l1.next
            l2 = l2.next
        
        # Process remaining l1
        while l1:
            
            value = l1.val + carry
            carry = value // 10
            
            l3.next = ListNode(value % 10)
            l3 = l3.next
            l1 = l1.next
        
        # Process remaining l2
        while l2:
            
            value = l2.val + carry
            carry = value // 10
            
            l3.next = ListNode(value % 10)
            l3 = l3.next
            l2 = l2.next
        
        # Add carry only if it exists
        if carry:
            l3.next = ListNode(carry)
        
        # Return result (skip dummy node)
        return head.next
```
