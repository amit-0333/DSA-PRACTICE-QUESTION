```python
# TC: O(n)       → we traverse the list once
# SC: O(1)       → in-place reversal, no extra space

def reverseList(head):
    
    # Step 1: Initialize pointers
    prev = None          # will become new head
    curr = head          # start from current head
    
    # Step 2: Traverse the list
    while curr:
        
        # Store next node before breaking link
        nxt = curr.next
        
        # Reverse the link
        curr.next = prev
        
        # Move pointers forward
        prev = curr
        curr = nxt
    
    # Step 3: prev is new head after reversal
    return prev
```
