```python
class Solution:
    
    # TC: O(N) 
    # Each node is visited once. Even though we traverse child lists,
    # overall traversal across all nodes is still linear.
    
    # SC: O(N)
    # Due to recursion stack in worst case (when every node has a child).
    
    def flatten(self, head):
        
        # Base case: if list is empty, return as it is
        if not head:
            return head
        
        # Start from head
        curr = head
        
        # Traverse the main list
        while curr:
            
            # If current node has a child list
            if curr.child:
                
                # Store next pointer to reconnect later
                nxt = curr.next
                
                # Recursively flatten the child list
                curr.next = self.flatten(curr.child)
                
                # Connect child list's head back to current node
                curr.next.prev = curr
                
                # Remove child pointer as it's now merged
                curr.child = None
                
                # Move to the end (tail) of the flattened child list
                temp = curr.next
                while temp.next:
                    temp = temp.next
                
                # Reconnect the original next part of the list
                if nxt:
                    temp.next = nxt
                    nxt.prev = temp
            
            # Move to next node in the list
            curr = curr.next
        
        # Return the flattened list head
        return head
```
