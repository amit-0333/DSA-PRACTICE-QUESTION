```python
# TC: O(n * m)   → merging multiple lists
# SC: O(n)       → recursion stack

class Solution:
    
    # Merge two sorted bottom/child lists
    def merge(self, a, b):
        
        # Base cases
        if not a:
            return b
        if not b:
            return a
        
        # Choose smaller node
        if a.data <= b.data:
            result = a
            result.bottom = self.merge(a.bottom, b)
        else:
            result = b
            result.bottom = self.merge(a, b.bottom)
        
        # Remove horizontal link
        result.next = None
        
        return result
    
    
    def flatten(self, root):
        
        # Base case
        if not root or not root.next:
            return root
        
        # Flatten right side
        root.next = self.flatten(root.next)
        
        # Merge current with right
        root = self.merge(root, root.next)
        
        return root
```
