```python
class Solution:
    
    # TC: O(N)
    # First pass → create all nodes and store in hashmap → O(N)
    # Second pass → assign random pointers → O(N)
    
    # SC: O(N)
    # Hashmap stores mapping of old nodes to new nodes
    
    def copyRandomList(self, head):
        
        # Base case: empty list
        if not head:
            return None
        
        # Hashmap to store mapping: old node -> new node
        h = {}
        
        # Create new head node
        newhead = Node(head.val)
        
        # Map original head to new head
        h[head] = newhead
        
        # Temporary pointers for traversal
        oldtemp = head.next
        newtemp = newhead
        
        # Step 1: Create all nodes and link next pointers
        while oldtemp:
            
            # Create copy of current node
            copynode = Node(oldtemp.val)
            
            # Store mapping
            h[oldtemp] = copynode
            
            # Link new list
            newtemp.next = copynode
            
            # Move both pointers
            oldtemp = oldtemp.next
            newtemp = newtemp.next
        
        # Step 2: Assign random pointers
        oldtemp = head
        newtemp = newhead
        
        while oldtemp:
            
            # If random exists, map it using hashmap
            if oldtemp.random:
                newtemp.random = h[oldtemp.random]
            else:
                newtemp.random = None
            
            # Move forward
            oldtemp = oldtemp.next
            newtemp = newtemp.next
        
        # Return head of copied list
        return newhead
```
