```python
class Solution:
    
    # TC: O(N)
    # We traverse the list once to reach 'left'
    # and once more to reverse the sublist → overall linear
    
    # SC: O(1)
    # No extra space used (in-place reversal)
    
    def reverseBetween(self, head, left: int, right: int):
        
        # Base case: empty list, single node, or no need to reverse
        if head is None or head.next is None or left == right:
            return head
        
        # Step 1: Reach the 'left' position
        i = 1
        curr = head
        prev = None
        
        while curr and i != left:
            prev = curr          # node before sublist
            curr = curr.next
            i += 1
        
        # pointerTostart → node just before sublist
        pointerTostart = prev
        
        # start → first node of sublist (will become tail after reverse)
        start = curr
        
        # Reset prev for reversal
        prev = None
        
        # Step 2: Reverse nodes from left to right
        nxt = None
        while curr and i != right + 1:
            
            nxt = curr.next      # store next node
            
            curr.next = prev     # reverse link
            
            prev = curr          # move prev forward
            curr = nxt           # move curr forward
            
            i += 1
        
        # Step 3: Reconnect reversed sublist
        
        # Connect tail of reversed part to remaining list
        start.next = curr
        
        # Connect first part to reversed sublist
        if pointerTostart:
            pointerTostart.next = prev
        else:
            # If left == 1, new head is 'prev'
            return prev
        
        return head
```
