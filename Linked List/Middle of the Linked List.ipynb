```python
# TC: O(n) + O(n) ≈ O(n)  → one pass for counting, one pass to reach middle
# SC: O(1)               → no extra space used

def middleNode(head):
    
    # Step 1: Count total nodes in linked list
    temp = head
    count = 0
    
    while temp:
        count += 1          # increment count
        temp = temp.next    # move to next node
    
    # Step 2: Find middle index
    mid = count // 2        # integer division
    
    # Step 3: Traverse again to reach middle node
    temp = head
    
    while mid > 0:
        temp = temp.next    # move forward
        mid -= 1            # decrease steps
    
    # Step 4: Return middle node
    return temp
```

```python
# TC: O(n)       → single traversal
# SC: O(1)       → no extra memory used

def middleNode(head):
    
    # Initialize two pointers
    slow = head   # moves 1 step
    fast = head   # moves 2 steps
    
    # Traverse until fast reaches end
    while fast and fast.next:
        slow = slow.next          # move by 1
        fast = fast.next.next     # move by 2
    
    # When fast reaches end, slow is at middle
    return slow
```
