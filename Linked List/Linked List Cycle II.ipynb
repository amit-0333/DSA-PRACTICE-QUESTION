```python
# TC: O(n)       → at most 2 traversals of the list
# SC: O(1)       → no extra space used (in-place)

def detectCycle(head):
    
    # Step 1: Initialize two pointers
    slow = head      # moves 1 step
    fast = head      # moves 2 steps
    
    # Step 2: Detect if cycle exists
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        # If both meet → cycle detected
        if slow == fast:
            break
    else:
        # No cycle
        return None
    
    # Step 3: Find start of cycle
    slow = head
    
    # Move both one step at a time
    # They will meet at cycle start
    while slow != fast:
        slow = slow.next
        fast = fast.next
    
    # Step 4: Return starting node of cycle
    return slow
```

```python
# TC: O(n)       → each node visited once
# SC: O(n)       → storing visited nodes in set

def detectCycle(head):
    
    seen = set()        # stores visited nodes
    temp = head         # start traversal
    
    while temp:
        
        # If node already visited → cycle detected
        if temp in seen:
            return temp     # start of cycle
        
        # Otherwise store node and move forward
        seen.add(temp)
        temp = temp.next
    
    # No cycle found
    return None
```
