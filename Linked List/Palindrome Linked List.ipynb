```python
class Solution:
    
    # TC: O(N)
    # First traversal to push all elements into stack → O(N)
    # Second traversal for comparison up to N/2 → O(N)
    # Overall: O(N)
    
    # SC: O(N)
    # Stack stores all N node values
    
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        
        # Base case:
        # Empty list or single node is always a palindrome
        if head is None or head.next is None:
            return True
        
        # Initialize stack to store values
        stack = []
        
        # Temporary pointer to traverse list
        temp = head
        
        # Step 1: Push all node values into stack
        while temp:
            stack.append(temp.val)   # store value
            temp = temp.next         # move forward
        
        # Reset temp to head for comparison
        temp = head
        
        # Step 2: Compare first half with reversed second half (via stack)
        for i in range(len(stack) // 2):
            
            # Pop gives last element (reverse order)
            stack_ele = stack.pop()
            
            # Compare with current node value
            if stack_ele == temp.val:
                
                # Move forward only if matched
                temp = temp.next
                continue
            
            else:
                # Mismatch → not a palindrome
                return False
        
        # All comparisons matched
        return True
```
