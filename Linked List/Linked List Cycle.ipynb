```python
# TC: O(n)       → at most one full traversal
# SC: O(1)       → no extra space used

def hasCycle(head):
    
    # Step 1: Initialize two pointers
    tortoise = head      # moves 1 step
    hare = head          # moves 2 steps
    
    # Step 2: Traverse the list
    while hare and hare.next:
        
        tortoise = tortoise.next       # move by 1
        hare = hare.next.next          # move by 2
        
        # Step 3: If they meet → cycle exists
        if tortoise == hare:
            return True
    
    # Step 4: If fast pointer reaches end → no cycle
    return False
```
