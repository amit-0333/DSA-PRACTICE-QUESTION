```python
class Solution:
    
    # TC: O(N)
    # One pass to calculate length + one pass to reach breaking point
    
    # SC: O(1)
    # No extra space used
    
    def rotateRight(self, hea, k: int):
        
        # Base case: empty list, single node, or no rotation needed
        if head is None or head.next is None or k == 0:
            return head
        
        # Step 1: Find length of list
        n = 1
        temp = head
        
        while temp.next:
            temp = temp.next
            n += 1
        
        # Step 2: Make list circular
        temp.next = head
        
        # Step 3: Reduce k (important optimization)
        k = k % n
        
        # If k becomes 0 → no change
        if k == 0:
            temp.next = None   # break circle
            return head
        
        # Step 4: Find new tail (n - k steps)
        jump = n - k
        temp = head
        
        while jump - 1:
            temp = temp.next
            jump -= 1
        
        # Step 5: Break the circle and set new head
        new_head = temp.next
        temp.next = None
        
        return new_head
```
