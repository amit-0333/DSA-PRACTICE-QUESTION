```python
class Solution:
    
    # TC: O(N)
    # Each node is visited once during reversal.
    # Recursive calls process disjoint segments → overall linear time.
    
    # SC: O(N / K)
    # Due to recursion stack (in worst case ~ N/K calls)
    
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        
        # Step 1: Check if at least k nodes exist
        count = 0
        temp = head
        
        while temp and count < k:
            temp = temp.next
            count += 1
        
        # If less than k nodes, no reversal needed
        if count < k:
            return head
        
        # Step 2: Reverse first k nodes
        curr = head
        prev = None
        nxt = None
        count = 0
        
        while curr and count < k:
            
            nxt = curr.next      # store next node
            
            curr.next = prev     # reverse link
            
            prev = curr          # move prev forward
            curr = nxt           # move curr forward
            
            count += 1
        
        # Step 3: Recursively process remaining list
        # 'head' is now the tail of reversed group
        if nxt:
            head.next = self.reverseKGroup(nxt, k)
        
        # Step 4: Return new head of this segment
        return prev
```
