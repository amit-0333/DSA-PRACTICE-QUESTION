```python
## Brute force
def subarraySumk(nums, k):
    # Time Complexity: O(n^2)
    # Reason: Two nested loops → for each starting index i,
    # we iterate through all possible ending indices j.

    # Space Complexity: O(1)
    # Reason: No extra space is used apart from a few variables.

    count = 0
    n = len(nums)

    # Outer loop → fixes the starting index of subarray
    for i in range(n):
        sum = 0  # stores sum of current subarray starting at i

        # Inner loop → extends subarray from i to j
        for j in range(i, n):
            sum += nums[j]  # add current element to sum

            # check if current subarray sum equals k
            if sum == k:
                count += 1  # found a valid subarray

    # return total number of subarrays with sum k
    return count


# Example usage
nums = [1, 1, 1]
k = 2
print(subarraySumk(nums, k))
```

```python
## Optimal solution
def subarraySum(nums, k):
    # Time Complexity: O(n)
    # Space Complexity: O(n)

    # freq dictionary stores count of prefix sums
    # {prefix_sum: frequency}
    freq = {0: 1}  # महत्वपूर्ण: handles subarrays starting from index 0

    curr_sum = 0   # running prefix sum
    count = 0      # number of subarrays with sum = k

    for num in nums:
        # add current element to prefix sum
        curr_sum += num

        # check if there exists a prefix sum such that:
        # curr_sum - previous_sum = k  => previous_sum = curr_sum - k
        val = curr_sum - k

        # if such prefix sum exists, add its frequency
        if val in freq:
            count += freq[val]

        # update frequency of current prefix sum
        freq[curr_sum] = freq.get(curr_sum, 0) + 1

    # return total count of valid subarrays
    return count

# Example usage
nums = [1, 1, 1]
k = 2
print(subarraySum(nums, k))
```
