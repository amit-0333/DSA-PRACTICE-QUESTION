```python
def containerWithMostWater(arr):
    # Time: O(n), Space: O(1)

    i = 0
    j = len(arr) - 1
    maxArea = 0

    while i < j:
        height = min(arr[i], arr[j])
        width = j - i
        maxArea = max(maxArea, height * width)

        if arr[i] < arr[j]:
            i += 1
        else:
            j -= 1

    return maxArea
```
