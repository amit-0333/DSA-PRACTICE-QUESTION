```python
def containerWithMostWater(arr):
    # Time Complexity: O(n)
    # Space Complexity: O(1)

    i = 0
    j = len(arr) - 1
    maxArea = 0

    while i < j:
        height = min(arr[i], arr[j])
        width = j - i
        maxArea = max(maxArea, height * width)

        # Move pointer of smaller height
        if arr[i] < arr[j]:
            i += 1
        else:
            j -= 1

    return maxArea


# Example
arr = [1,8,6,2,5,4,8,3,7]
print(containerWithMostWater(arr))
```
