```python
## Brute force
# Time Complexity (TC): O(n^2)
# Space Complexity (SC): O(1) (excluding output)

def productExceptSelf(nums):
    n = len(nums)
    ans = [1] * n   

    for i in range(n):            # TC: O(n^2)
        for j in range(n):
            if i != j:
                ans[i] *= nums[j]   
    
    return ans


nums = [1,2,3,4]
print(productExceptSelf(nums))
```

```python
# better solution
# Overall Time Complexity (TC): O(n)
# Space Complexity (SC): O(n)  (prefix + suffix arrays)
def productExceptSelf(nums):
    n = len(nums)
    
    prefix = [1] * n
    suffix = [1] * n
    ans = [1] * n

    # Build prefix
    # TC: O(n)
    for i in range(1, n):
        prefix[i] = prefix[i-1] * nums[i-1]

    # Build suffix
    # TC: O(n)
    for i in range(n-2, -1, -1):
        suffix[i] = suffix[i+1] * nums[i+1]

    # Build answer
    # TC: O(n)
    for i in range(n):
        ans[i] = prefix[i] * suffix[i]

    return ans


nums = [1,2,3,4]
print(productExceptSelf(nums))
```

```python
# optimal solution
# Overall Time Complexity (TC): O(n)
# Space Complexity (SC): O(1)  (excluding output array)
def productExceptSelf(nums):
    n = len(nums)
    ans = [1] * n 

    # Step 1: Build prefix (left product)
    # TC: O(n)
    for i in range(1, n):
        ans[i] = ans[i-1] * nums[i-1]

    # Step 2: Multiply with suffix (right product)
    # TC: O(n)
    suffix = 1
    for i in range(n-2, -1, -1):
        suffix *= nums[i+1]
        ans[i] *= suffix

    return ans

nums = [1,2,3,4]
print(productExceptSelf(nums))
```
