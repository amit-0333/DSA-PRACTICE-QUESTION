```python
def compress(chars):
    # Time Complexity: O(n)
    # - each character visited once

    # Space Complexity: O(1)
    # - in-place modification

    idx = 0   # write pointer
    i = 0     # read pointer
    n = len(chars)

    while i < n:
        ch = chars[i]
        count = 0

        # count same characters
        while i < n and chars[i] == ch:
            i += 1
            count += 1

        # write character
        if count==1:
            chars[idx] = ch
            idx += 1

        # write count if > 1
        if count > 1:
            for dig in str(count):
                chars[idx] = dig
                idx += 1

    return idx
```
