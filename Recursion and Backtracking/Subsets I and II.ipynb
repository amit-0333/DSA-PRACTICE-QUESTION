```python
# Time Complexity (TC): O(2^n)
# Reason: For each element, we have 2 choices (take / not take), leading to 2^n subsets.
#
# Space Complexity (SC): O(n)
# Reason: Recursion depth is n (call stack) and temporary list "ans" can hold up to n elements.

def solve(arr, i, ans, allsubset):
    # Base condition:
    # If we have considered all elements, store the current subset
    if i == len(arr):
        allsubset.append(ans.copy())  # store a copy to avoid reference issues
        return

    # Choice 1: Include current element
    ans.append(arr[i])
    solve(arr, i + 1, ans, allsubset)

    # Backtrack: remove last element before exploring next choice
    ans.pop()

    # Choice 2: Exclude current element
    solve(arr, i + 1, ans, allsubset)


# Time Complexity (TC): O(2^n)
# Space Complexity (SC): O(n)
# Reason: Same as above + recursion stack

def subsets1(arr):
    allsubset = []
    solve(arr, 0, [], allsubset)
    return allsubset

# driver
nums = [1, 2, 2]
print(subsets1(nums))
```

```python
# Time Complexity (TC): O(2^n)
# Reason: In worst case, each element has 2 choices (pick / not pick),
# but duplicates reduce some branches; still exponential overall.
#
# Space Complexity (SC): O(n)
# Reason: Recursion depth is n, and temporary list "ans" stores at most n elements.

def solve(arr, i, ans, res):
    # Base case:
    # If we have processed all elements, store current subset
    if i == len(arr):
        res.append(ans[:])  # copy to avoid reference issues
        return

    # Choice 1: Pick element
    ans.append(arr[i])
    solve(arr, i + 1, ans, res)

    # Backtrack: remove last added element
    ans.pop()


    # Skip duplicates 
    # If next elements are same, move i forward to avoid duplicate subsets
    while i + 1 < len(arr) and arr[i] == arr[i + 1]:
        i += 1

    # Choice 2: Not pick element
    solve(arr, i + 1, ans, res)


# Time Complexity (TC): O(n log n + 2^n)
# Reason: Sorting takes O(n log n), recursion generates subsets.
#
# Space Complexity (SC): O(n)
# Reason: Recursion stack + temporary subset storage.

def subset2(nums):
    nums.sort()  # sort to handle duplicates easily
    res = []
    solve(nums, 0, [], res)
    return res


# Driver code
nums = [1, 2, 2]
print(subset2(nums))
```
