```python
# Time Complexity: O(N!)
# Space Complexity: O(N) recursion stack (excluding output)

def isSafe(board, row, col, n):
    
    # check entire row
    for j in range(n):
        if board[row][j] == 'Q':
            return False

    # check entire column
    for i in range(n):
        if board[i][col] == 'Q':
            return False
    
    # check top-left diagonal
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 'Q':
            return False
        i -= 1
        j -= 1
    
    # check top-right diagonal
    i, j = row, col
    while i >= 0 and j < n:
        if board[i][j] == 'Q':
            return False
        i -= 1
        j += 1
    
    return True


def nQueens(board, row, n, ans):
    
    # base case: all queens placed
    if row == n:
        temp = ["".join(r) for r in board]  # convert row list → string
        ans.append(temp)
        return
    
    # try placing queen in every column of current row
    for col in range(n):
        
        # check if position is safe
        if isSafe(board, row, col, n):
            
            board[row][col] = 'Q'   # place queen
            
            nQueens(board, row+1, n, ans)  # move to next row
            
            board[row][col] = '.'   # backtrack (remove queen)


def solveNQueens(n):
    
    # create empty board
    board = [['.' for _ in range(n)] for _ in range(n)]
    ans = []
    
    # start from row 0
    nQueens(board, 0, n, ans)
    
    return ans

n=4
print(solveNQueens(n))
```
