```python
# Time Complexity (TC): O(n log n)
# Reason: Array is divided into halves (log n levels),
# and each level takes O(n) time to merge.
#
# Space Complexity (SC): O(n)
# Reason: Temporary arrays are used during merge + recursion stack O(log n)


def mergeProcedure(nums, i, mid, j):
    # Size of left and right subarrays
    nl = mid - i + 1
    nr = j - mid

    # Temporary arrays
    ls = [0] * nl
    rs = [0] * nr

    # Copy elements into left subarray
    for m in range(nl):
        ls[m] = nums[i + m]

    # Copy elements into right subarray
    for n in range(nr):
        rs[n] = nums[mid + 1 + n]

    # Pointers for merging
    p = 0
    q = 0
    k = i

    # Merge both arrays in sorted order
    while p < nl and q < nr:
        if ls[p] <= rs[q]:
            nums[k] = ls[p]
            p += 1
        else:
            nums[k] = rs[q]
            q += 1
        k += 1

    # Copy remaining elements of left array
    while p < nl:
        nums[k] = ls[p]
        p += 1
        k += 1

    # Copy remaining elements of right array
    while q < nr:
        nums[k] = rs[q]
        q += 1
        k += 1


# Time Complexity (TC): O(n log n)
# Reason: Each recursive level splits array into halves and merging takes O(n)
#
# Space Complexity (SC): O(log n)
# Reason: Recursion stack depth is log n


def mergeSort(nums, i, j):
    # Base case: single element
    if i >= j:
        return

    # Find middle point
    mid = i + (j - i) // 2

    # Sort left half
    mergeSort(nums, i, mid)

    # Sort right half
    mergeSort(nums, mid + 1, j)

    # Merge sorted halves
    mergeProcedure(nums, i, mid, j)


# Time Complexity (TC): O(n log n)
# Space Complexity (SC): O(log n)

def sortArray(nums):
    mergeSort(nums, 0, len(nums) - 1)
    return nums


# Driver code
nums = [5, 2, 3, 1]
print(sortArray(nums))
```
