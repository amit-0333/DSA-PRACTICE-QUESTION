```python
# Time Complexity (TC): O(4^(n^2))
# Reason: Each cell has 4 possible directions, worst-case explores all paths.
#
# Space Complexity (SC): O(n^2)
# Reason: Visited markings + recursion stack depth

def dfs(i, j, maze, n):
    # If reached destination
    if i == n - 1 and j == n - 1:
        return True

    # Boundary + invalid cell check
    if i < 0 or i >= n or j < 0 or j >= n or maze[i][j] == 0:
        return False

    # Mark visited
    temp = maze[i][j]
    maze[i][j] = 0

    # Explore all 4 directions
    found = (
        dfs(i + 1, j, maze, n) or
        dfs(i - 1, j, maze, n) or
        dfs(i, j - 1, maze, n) or
        dfs(i, j + 1, maze, n)
    )

    # Backtrack
    maze[i][j] = temp

    return found


def findPath(m, n):
    # If starting cell is blocked
    if m[0][0] == 0:
        return False

    return dfs(0, 0, m, n)

# Driver Code
maze = [
    [1, 0, 0, 0],
    [1, 1, 0, 1],
    [0, 1, 0, 0],
    [1, 1, 1, 1]
]

n = len(maze)

if findPath(maze, n):
    print("Path exists")
else:
    print("No path exists")
```
