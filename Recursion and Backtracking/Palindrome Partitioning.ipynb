```python
# Time Complexity: O(n * 2^n)
# - There are ~2^n possible partitions
# - For each substring, palindrome check takes O(n)

# Space Complexity: 
# - O(n) recursion stack (depth of recursion)
# - O(n) for current combination
# - O(n * 2^n) to store all answers (output)

def isPalindrome(sub):
    # check if string is palindrome
    return sub == sub[::-1]


def palpart(s, i, comb, ans):
    
    # Base Case: if we reach end of string
    if i == len(s):
        ans.append(comb.copy())
        return
    
    # Try all substrings starting from index i
    for j in range(i, len(s)):
        
        # take substring
        sub = s[i:j+1]
        
        # check palindrome
        if isPalindrome(sub):
            
            # choose
            comb.append(sub)
            
            # explore
            palpart(s, j+1, comb, ans)
            
            # backtrack
            comb.pop()


def partition(s):
    ans = []
    palpart(s, 0, [], ans)
    return ans

# driver code

s = "aab"

result = partition(s)

print(result)
```
