```python
# Time Complexity: exponential in worst case (pruned heavily)
# Space Complexity: O(n^2) recursion depth


def isValid(board, r, c, n, expmove):
    
    # invalid case: out of bounds OR wrong move number
    if r < 0 or r >= n or c < 0 or c >= n or board[r][c] != expmove:
        return False
    
    # if last move reached
    if expmove == n*n - 1:
        return True
    
    # check all 8 knight moves
    return (
        isValid(board, r-2, c+1, n, expmove+1) or
        isValid(board, r-1, c+2, n, expmove+1) or
        isValid(board, r+1, c+2, n, expmove+1) or
        isValid(board, r+2, c+1, n, expmove+1) or
        isValid(board, r+2, c-1, n, expmove+1) or
        isValid(board, r+1, c-2, n, expmove+1) or
        isValid(board, r-1, c-2, n, expmove+1) or
        isValid(board, r-2, c-1, n, expmove+1)
    )


def checkValidGrid(grid):
    
    n = len(grid)
    
    # must start from 0 at (0,0)
    if grid[0][0] != 0:
        return False
    
    return isValid(grid, 0, 0, n, 0)

grid = [[0,11,16,5,20],[17,4,19,10,15],[12,1,8,21,6],[3,18,23,14,9],[24,13,2,7,22]]
print(checkValidGrid(grid))
```
