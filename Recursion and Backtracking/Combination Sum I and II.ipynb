```python
# Time Complexity: O(2^n * k)
# Each element has two choices (take or skip), forming a recursion tree of size ~2^n.
# For every valid combination, copying takes O(k), where k is length of combination.

# Space Complexity: O(k)
# Recursion depth can go up to k due to repeated inclusion of same element.
# Additional space used for storing current combination.

def cs(candidates, target, i, n, comb, ans):
    
    # If target becomes 0, store the current combination
    if target == 0:
        ans.append(comb.copy())
        return
    
    # If index goes out of range or target becomes negative, stop
    if i == n or target < 0:
        return
    
    # Take current element
    comb.append(candidates[i])
    # Stay at same index because element can be reused
    cs(candidates, target - candidates[i], i, n, comb, ans)
    
    # Backtrack
    comb.pop()
    
    # Skip current element and move to next index
    cs(candidates, target, i + 1, n, comb, ans)


def combinationSum(candidates, target):
    ans = []
    comb = []
    n = len(candidates)
    
    cs(candidates, target, 0, n, comb, ans)
    
    return ans

candidates = [2, 3, 6, 7]
target = 7

result = combinationSum(candidates, target)

print(result)
```

```python
# Time Complexity: O(2^n * k)
# We explore all subsets using take/skip decisions, giving ~2^n possibilities.
# For each valid combination, copying takes O(k).

# Space Complexity: O(k)
# Recursion depth can go up to k.
# Extra space is used for current combination.


def cs2(candidates, target, i, n, comb, ans):
    # This function generates all unique combinations starting from index i.
    # At each index, we either take the element or skip it.
    # Since each number can be used only once, after taking we move to i+1.
    # To avoid duplicates, we skip all equal elements before making the skip call.
    
    if target == 0:
        # If target becomes zero, current combination is valid
        ans.append(comb.copy())
        return
    
    if i == n or target < 0:
        # If we reach end or target goes negative, stop
        return
    
    # Take current element
    comb.append(candidates[i])
    cs2(candidates, target - candidates[i], i + 1, n, comb, ans)
    
    # Backtrack
    comb.pop()
    
    # Skip duplicates of current element
    while i + 1 < n and candidates[i] == candidates[i + 1]:
        i += 1
    
    # Skip current element and move to next unique element
    cs2(candidates, target, i + 1, n, comb, ans)


def combinationSum2(candidates: List[int], target: int) -> List[List[int]]:
    # This function prepares data and starts recursion.
    # Sorting is required so duplicates come together,
    # which helps in skipping them properly.
    
    candidates.sort()
    
    ans = []
    comb = []
    n = len(candidates)
    
    cs2(candidates, target, 0, n, comb, ans)
    
    return ans


candidates = [10, 1, 2, 7, 6, 1, 5]
target = 8

result = combinationSum2(candidates, target)

print(result)
```
