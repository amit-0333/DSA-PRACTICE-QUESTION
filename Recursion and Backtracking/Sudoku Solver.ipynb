```python
# Time Complexity: ~O(9^(empty cells))  (backtracking tries 1–9 per empty cell)
# Space Complexity: O(1) extra (board modified in-place) + O(depth) recursion stack

def isSafe(board, row, col, dig):
    """
    Check if placing 'dig' at (row, col) is valid.
    We verify:
    1) Same row
    2) Same column
    3) Same 3x3 sub-grid
    """
    
    # check row: no same digit in this row
    for j in range(9):
        if board[row][j] == dig:
            return False
    
    # check column: no same digit in this column
    for i in range(9):
        if board[i][col] == dig:
            return False
    
    # check 3x3 box:
    # find top-left corner of the 3x3 box containing (row, col)
    sr = (row // 3) * 3
    sc = (col // 3) * 3
    
    for i in range(sr, sr + 3):
        for j in range(sc, sc + 3):
            if board[i][j] == dig:
                return False
    
    # if all checks pass, it's safe to place
    return True


def solver(board, row, col):
    """
    Fill the board using backtracking, moving cell by cell.
    Traversal order: left → right, then next row.
    """
    
    # base case:
    # if we have reached beyond the last row, puzzle is solved
    if row == 9:
        return True
    
    # compute next cell position
    nxtrow = row
    nxtcol = col + 1
    
    # if we reach end of column, move to next row
    if nxtcol == 9:
        nxtrow = row + 1
        nxtcol = 0
    
    # if current cell already filled, skip to next cell
    # IMPORTANT: return directly to avoid extra recursion
    if board[row][col] != '.':
        return solver(board, nxtrow, nxtcol)
    
    # try placing digits 1 to 9 in current empty cell
    for d in range(1, 10):
        dig = str(d)
        
        # check if placing 'dig' here is valid
        if isSafe(board, row, col, dig):
            
            # place digit
            board[row][col] = dig
            
            # move to next cell
            if solver(board, nxtrow, nxtcol):
                return True   # solution found, stop further work
            
            # backtrack: undo the placement
            board[row][col] = '.'
    
    # if no digit works here, return False to backtrack
    return False


def solveSudoku(board):
    """
    Driver function: modifies board in-place to a valid solution.
    """
    solver(board, 0, 0)

# example Sudoku board ('.' means empty)

board = [
    ["5","3",".",".","7",".",".",".","."],
    ["6",".",".","1","9","5",".",".","."],
    [".","9","8",".",".",".",".","6","."],
    ["8",".",".",".","6",".",".",".","3"],
    ["4",".",".","8",".","3",".",".","1"],
    ["7",".",".",".","2",".",".",".","6"],
    [".","6",".",".",".",".","2","8","."],
    [".",".",".","4","1","9",".",".","5"],
    [".",".",".",".","8",".",".","7","9"]
]

# call solver
solveSudoku(board)

# print solved board
for row in board:
    print(" ".join(row))

```
