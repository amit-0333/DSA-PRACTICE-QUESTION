```python
# Time Complexity: O(M^N)
# In worst case, for each of N nodes we try M colors → M * M * ... (N times)

# Space Complexity: O(N)
# - colors array takes O(N)
# - recursion stack depth is also O(N)


def isSafe(graph, node, colors, col):
    """
    Check if we can assign color 'col' to current node.
    
    graph[node] → gives all neighbors of 'node'
    We must ensure no neighbor has same color.
    """
    
    for nei in graph[node]:        # check all adjacent nodes
        if colors[nei] == col:     # if neighbor already has same color
            return False           # not safe
    
    return True                    # safe to assign


def solve(graph, node, m, colors):
    """
    Recursive function to try coloring each node.
    
    node → current node we are trying to color
    m → number of colors available
    colors → array storing assigned colors
    """
    
    n = len(graph)
    
    # base case: if all nodes are colored
    if node == n:
        return True               # valid coloring found
    
    # try assigning each color from 1 to m
    for col in range(1, m + 1):
        
        # check if current color can be assigned
        if isSafe(graph, node, colors, col):
            
            colors[node] = col    # assign color
            
            # move to next node
            if solve(graph, node + 1, m, colors):
                return True       # if solution found, stop
            
            colors[node] = 0      # backtrack (remove color)
    
    # if no color works for this node
    return False


def graphColoring(graph, m, n):
    """
    Driver function
    
    graph → adjacency list
    m → number of colors
    n → number of nodes
    """
    
    colors = [0] * n   # initially all nodes uncolored (0)
    
    return solve(graph, 0, m, colors)

# example graph (adjacency list)
graph = [
    [1, 2],   # node 0 connected to 1, 2
    [0, 2],   # node 1 connected to 0, 2
    [0, 1]    # node 2 connected to 0, 1
]

n = 3   # number of nodes
m = 3   # number of colors

# call function
result = graphColoring(graph, m, n)

print(result)
```
