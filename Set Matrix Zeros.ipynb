```python
# Overall Time Complexity (TC): O(n*m)
# Space Complexity (SC): O(n + m)

def setZeroes(matrix):
    n = len(matrix)
    m = len(matrix[0])

    row_zero = set()
    col_zero = set()

    # Step 1: find rows & columns having 0
    # TC: O(n*m)
    for i in range(n):
        for j in range(m):
            if matrix[i][j] == 0:
                row_zero.add(i)
                col_zero.add(j)

    # Step 2: update matrix
    # TC: O(n*m)
    for i in range(n):
        for j in range(m):
            if i in row_zero or j in col_zero:  # ✅ OR condition
                matrix[i][j] = 0

    return matrix

matrix = [
    [1, 1, 1],
    [1, 0, 1],
    [1, 1, 1]
]

setZeroes(matrix)

print(matrix)
```

```python
# optimal solution
# Overall Time Complexity (TC): O(n*m)
# Space Complexity (SC): O(1)

def setZeroes(matrix):
    n = len(matrix)
    m = len(matrix[0])

    first_row_zero = False
    first_col_zero = False

    # Step 1: check first row
    for j in range(m):              # TC: O(m)
        if matrix[0][j] == 0:
            first_row_zero = True

    # Step 2: check first column
    for i in range(n):              # TC: O(n)
        if matrix[i][0] == 0:
            first_col_zero = True

    # Step 3: mark rows & columns
    for i in range(1, n):           # TC: O(n*m)
        for j in range(1, m):
            if matrix[i][j] == 0:
                matrix[0][j] = 0
                matrix[i][0] = 0

    # Step 4: update matrix using markers
    for i in range(1, n):           # TC: O(n*m)
        for j in range(1, m):
            if matrix[0][j] == 0 or matrix[i][0] == 0:
                matrix[i][j] = 0

    # Step 5: update first row
    if first_row_zero:              # TC: O(m)
        for j in range(m):
            matrix[0][j] = 0

    # Step 6: update first column
    if first_col_zero:              # TC: O(n)
        for i in range(n):
            matrix[i][0] = 0


matrix = [
    [1, 1, 1],
    [1, 0, 1],
    [1, 1, 1]
]

setZeroes(matrix)

print(matrix)
```
