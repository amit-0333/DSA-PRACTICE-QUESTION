```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Order Agnostic Binary Search (works for both increasing & decreasing)
def orderAgnosticSearch(arr, target):
    if not arr:   # handle empty array
        return -1

    i = 0
    j = len(arr) - 1

    # check if array is increasing or decreasing
    isAscending = arr[0] < arr[-1]

    while i <= j:
        mid = i + (j - i) // 2   # avoids overflow

        if arr[mid] == target:
            return mid

        if isAscending:
            # ascending order logic
            if arr[mid] < target:
                i = mid + 1
            else:
                j = mid - 1
        else:
            # descending order logic
            if arr[mid] > target:
                i = mid + 1
            else:
                j = mid - 1

    return -1   # target not found


# test
arr = [9, 7, 5, 3, 1]
target = 9
print(orderAgnosticSearch(arr, target))
```
