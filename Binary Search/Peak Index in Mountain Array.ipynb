```python
# Time Complexity: O(log n)
# Space Complexity: O(1)
def peakIndexInMountainArray(arr):
    # i -> start pointer
    # j -> end pointer
    i = 0
    j = len(arr) - 1
    n = len(arr)

    # Binary Search
    while i <= j:
        mid = i + (j - i) // 2  # find middle index

        # Case 1: mid is not at boundary
        if mid > 0 and mid < n - 1:

            # If mid is greater than both neighbors → peak found
            if arr[mid] >= arr[mid - 1] and arr[mid] >= arr[mid + 1]:
                return mid

            # If right neighbor is greater → move right (increasing slope)
            elif arr[mid + 1] > arr[mid]:
                i = mid + 1

            # Else move left (decreasing slope)
            else:
                j = mid - 1

        # Case 2: handle boundary (edge cases)
        else:
            # If mid is first index
            if mid == 0:
                if arr[0] > arr[1]:
                    return 0
                else:
                    return 1

            # If mid is last index
            if mid == n - 1:
                if arr[n - 1] > arr[n - 2]:
                    return n - 1
                else:
                    return n - 2

    # If no peak found (should not happen in valid mountain array)
    return -1

arr = [0,10,5,2]
print(peakIndexInMountainArray(arr))
```

```python
# Time Complexity: O(log n)
# Space Complexity: O(1)
def peakIndexInMountainArray(arr):
    # i -> start pointer
    # j -> end pointer
    i = 0
    j = len(arr) - 1

    # Binary Search
    while i < j:
        mid = i + (j - i) // 2  # avoid overflow

        # If mid element is smaller than next,
        # we are on the increasing slope
        if arr[mid] < arr[mid + 1]:
            i = mid + 1   # move right
        else:
            # we are on decreasing slope OR at peak
            j = mid       # keep mid (could be peak)

    # when loop ends, i == j -> peak index
    return i

arr = [0,10,5,2]
print(peakIndexInMountainArray(arr))
```
