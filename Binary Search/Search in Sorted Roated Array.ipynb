```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Find index of minimum element (pivot)
def sortedRotatedArr(arr):
    i = 0
    j = len(arr) - 1
    n = len(arr)

    while i <= j:
        if arr[i] <= arr[j]:
            return i

        mid = i + (j - i) // 2

        nxt = (mid + 1) % n
        prev = (mid + n - 1) % n

        if arr[mid] <= arr[prev] and arr[mid] <= arr[nxt]:
            return mid
        elif arr[i] <= arr[mid]:
            i = mid + 1
        else:
            j = mid - 1


# Binary Search with given range
def binarySearch(arr, ele, i, j):
    while i <= j:
        mid = i + (j - i) // 2

        if arr[mid] == ele:
            return mid
        elif arr[mid] > ele:
            j = mid - 1
        else:
            i = mid + 1

    return -1


# Search in rotated sorted array
def search(arr, ele):
    n = len(arr)
    minEle = sortedRotatedArr(arr)

    # search in right sorted part
    if arr[minEle] <= ele <= arr[n - 1]:
        return binarySearch(arr, ele, minEle, n - 1)
    else:
        return binarySearch(arr, ele, 0, minEle - 1)


# Driver code
arr = [4, 5, 6, 7, 0, 1, 2]
ele = 1

print("Index:", search(arr, ele))
```
