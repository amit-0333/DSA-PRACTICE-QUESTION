```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Find index of minimum element in sorted rotated array
def sortedRotatedArr(arr):
    i = 0
    j = len(arr) - 1
    n = len(arr)

    while i <= j:
        # if subarray is already sorted
        if arr[i] <= arr[j]:
            return i

        mid = i + (j - i) // 2

        nxt = (mid + 1) % n
        prev = (mid + n - 1) % n

        # check if mid is minimum
        if arr[mid] <= arr[prev] and arr[mid] <= arr[nxt]:
            return mid

        # left part sorted → go right
        elif arr[i] <= arr[mid]:
            i = mid + 1
        else:
            j = mid - 1


# Driver code
arr = [4, 5, 6, 7, 0, 1, 2]

index = sortedRotatedArr(arr)

print("Index of minimum element:", index)
print("Minimum element:", arr[index])
print("Number of rotations:", index)
```
