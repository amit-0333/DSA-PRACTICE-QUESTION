```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Standard Binary Search
def binarySearch(arr, key, i, j):
    while i <= j:
        mid = i + (j - i) // 2

        if arr[mid] == key:
            return mid
        elif arr[mid] > key:
            j = mid - 1
        else:
            i = mid + 1

    return -1


# Find index of minimum element (pivot)
def rotatedArr(arr):
    i = 0
    j = len(arr) - 1
    n = len(arr)

    # already sorted
    if arr[i] <= arr[j]:
        return 0

    while i <= j:
        mid = i + (j - i) // 2

        nxt = (mid + 1) % n
        prev = (mid + n - 1) % n

        # check if mid is pivot
        if arr[mid] < arr[prev] and arr[mid] < arr[nxt]:
            return mid

        # fixed reference
        elif arr[mid] >= arr[0]:
            i = mid + 1
        else:
            j = mid - 1

    return 0


def search(nums, target):
    arr = nums
    n = len(arr)

    # single element case
    if n == 1:
        return 0 if arr[0] == target else -1

    # find pivot
    idx = rotatedArr(arr)

    # if not rotated
    if idx == 0:
        return binarySearch(arr, target, 0, n - 1)

    # choose correct half
    if arr[idx] <= target <= arr[n - 1]:
        return binarySearch(arr, target, idx, n - 1)
    else:
        return binarySearch(arr, target, 0, idx - 1)

nums = [4,5,6,7,0,1,2]
target = 0
print(search(nums,target))
```
