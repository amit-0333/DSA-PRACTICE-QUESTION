```python

# Binary Search on decreasing sorted array
#  Time Complexity: O(log n)
#  Space Complexity: O(1)
def decsSortedarr(arr, target):
    i = 0
    j = len(arr) - 1

    while i <= j:
        mid = i + (j - i) // 2   # avoid overflow

        if arr[mid] == target:
            return mid

        elif arr[mid] > target:
            i = mid + 1   # move right (since array is decreasing)

        else:
            j = mid - 1   # move left

    return -1   # target not found


arr=[9, 7, 5, 3, 1]
target=9
print(decsSortedarr(arr, target))
```
