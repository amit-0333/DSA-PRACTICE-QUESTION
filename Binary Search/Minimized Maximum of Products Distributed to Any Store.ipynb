```python
def isValid(quantities, n, mid):
    stores = 0   # count how many stores are needed

    for q in quantities:
        # number of stores needed for this product
        # ceil(q / mid)
        stores += (q + mid - 1) // mid

    # valid if total stores needed is within limit
    return stores <= n


def minimizedMaximum(n,quantities):
    i = 1                      # minimum possible max (at least 1 product per store)
    j = sum(quantities)       # maximum possible (all products in one store)
    res = j                   # store answer

    # binary search on answer
    while i <= j:
        mid = i + (j - i) // 2   # candidate maximum per store

        if isValid(quantities, n, mid):
            # if possible, try to minimize further
            res = mid
            j = mid - 1
        else:
            # not possible → need larger max
            i = mid + 1

    return res
    
quantities = [11,6]
n = 6
print(minimizedMaximum(n,quantities))
```
