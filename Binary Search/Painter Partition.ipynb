```python
# Time Complexity: O(n log(sum))
#   Binary search runs log(sum(arr)) times
#   Each check (isPossible) takes O(n)

# Space Complexity: O(1)
#   No extra space used

def isPossible(arr, k, mid):
    n = len(arr)
    painter = 1       # start with first painter
    time = 0          # current workload

    for i in range(n):
        # if current board can be painted within limit
        if time + arr[i] <= mid:
            time += arr[i]
        else:
            # assign new painter
            painter += 1
            time = arr[i]

            # if painters exceed k → not possible
            if painter > k:
                return False

    # all boards painted within k painters
    return True


def painterPartition(arr, k):
    i = max(arr)      # minimum possible answer
    j = sum(arr)      # maximum possible answer
    ans = -1

    while i <= j:
        mid = i + (j - i) // 2

        if isPossible(arr, k, mid):
            ans = mid
            j = mid - 1   # try smaller answer
        else:
            i = mid + 1   # increase limit

    return ans

arr = [10, 20, 30, 40]
k = 2

result = painterPartition(arr, k)
print("Minimum time:", result)
```
