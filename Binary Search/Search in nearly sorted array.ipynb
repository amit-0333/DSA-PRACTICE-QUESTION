```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Search in nearly sorted array
def search(arr, key):
    i = 0
    j = len(arr) - 1

    while i <= j:
        mid = i + (j - i) // 2

        if arr[mid] == key:
            return mid

        # check left neighbor
        if mid - 1 >= i and arr[mid - 1] == key:
            return mid - 1

        # check right neighbor
        if mid + 1 <= j and arr[mid + 1] == key:
            return mid + 1

        # move search space
        elif arr[mid] > key:
            j = mid - 2
        else:
            i = mid + 2

    return -1   # not found


# Driver code
arr = [5, 10, 30, 20, 40]
key = 20

print(search(arr, key))
```
