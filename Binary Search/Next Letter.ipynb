```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Find index of next greater letter (wrap-around allowed)
def nxtLetter(arr, letter):
    i = 0
    j = len(arr) - 1
    res = 0   # default for wrap-around

    while i <= j:
        mid = i + (j - i) // 2

        if arr[mid] > letter:
            res = mid        # possible answer
            j = mid - 1      # try smaller index
        else:
            i = mid + 1

    return res


# Driver code
arr = ['a', 'c', 'f', 'h']
letter = 'f'

idx = nxtLetter(arr, letter)
print("Index:", idx)
print("Next letter:", arr[idx])
```
