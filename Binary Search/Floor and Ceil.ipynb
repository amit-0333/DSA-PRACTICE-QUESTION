```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Find floor of element (greatest element <= ele)
def floorOfele(arr, ele):
    i = 0
    j = len(arr) - 1
    res = -1   # stores floor value

    while i <= j:
        mid = i + (j - i) // 2

        if arr[mid] == ele:
            return arr[mid]   # exact match → floor is itself

        elif arr[mid] < ele:
            res = arr[mid]    # possible floor
            i = mid + 1       # try to find larger one

        else:
            j = mid - 1

    return res

# Driver code
arr = [1, 2, 4, 6, 8]
ele = 5

print("Floor value:", floorOfele(arr, ele))
```

```python
## Gives Index
# Time Complexity: O(log n)
# Space Complexity: O(1)
def floorIndex(arr, ele):
    i = 0
    j = len(arr) - 1
    res = -1   # stores index

    while i <= j:
        mid = i + (j - i) // 2

        if arr[mid] == ele:
            return mid

        elif arr[mid] < ele:
            res = mid
            i = mid + 1
        else:
            j = mid - 1

    return res

# Driver code
arr = [1, 2, 4, 6, 8]
ele = 5

idx = floorIndex(arr, ele)

if idx != -1:
    print("Floor index:", idx)
    print("Floor value:", arr[idx])
else:
    print("No floor found")
```

```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Find ceil of element (smallest element >= ele)
def CeilOfele(arr, ele):
    i = 0
    j = len(arr) - 1
    res = -1   # stores ceil value

    while i <= j:
        mid = i + (j - i) // 2

        if arr[mid] == ele:
            return arr[mid]   # exact match → ceil is itself

        elif arr[mid] > ele:
            res = arr[mid]    # possible ceil
            j = mid - 1       # try to find smaller valid ceil

        else:
            i = mid + 1

    return res


# Driver code
arr = [1, 2, 4, 6, 8]
ele = 5

print("Ceil value:", CeilOfele(arr, ele))
```
