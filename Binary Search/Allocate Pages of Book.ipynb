```python
# Time Complexity: O(n log(sum(arr)))
# Space Complexity: O(1)

def isValid(arr, n, mid):
    students = 1
    curr_sum = 0

    for i in range(len(arr)):
        # if a single book is greater than mid → impossible
        if arr[i] > mid:
            return False

        curr_sum += arr[i]

        # if limit exceeded → assign to next student
        if curr_sum > mid:
            students += 1
            curr_sum = arr[i]

        # if students exceed allowed
        if students > n:
            return False

    return True


def allocateBooks(arr, n):
    i = 0
    j = sum(arr)   # maximum possible
    res = -1

    while i <= j:
        mid = i + (j - i) // 2

        if isValid(arr, n, mid):
            res = mid
            j = mid - 1
        else:
            i = mid + 1

    return res

arr=[10,20,30,40]
n=2
print(allocateBooks(arr,n))
```
