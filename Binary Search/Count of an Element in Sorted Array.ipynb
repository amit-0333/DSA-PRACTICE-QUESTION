```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Find first occurrence of key in sorted array
def firstOccurance(arr, key):
    i = 0
    j = len(arr) - 1
    res = -1   # store answer

    while i <= j:
        mid = i + (j - i) // 2

        if arr[mid] == key:
            res = mid        # store index
            j = mid - 1      # move left to find earlier occurrence

        elif arr[mid] > key:
            j = mid - 1

        else:
            i = mid + 1

    return res   # return stored result


# test
arr = [1, 2, 2, 2, 3, 4]
key = 2
print(firstOccurance(arr, key))
```

```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Find last occurrence of key in sorted array
def lastOccurance(arr, key):
    i = 0
    j = len(arr) - 1
    res = -1   # store answer

    while i <= j:
        mid = i + (j - i) // 2

        if arr[mid] == key:
            res = mid        # store index
            i = mid + 1      # move right to find later occurrence

        elif arr[mid] > key:
            j = mid - 1

        else:
            i = mid + 1

    return res   # return stored result


# test
arr = [1, 2, 2, 2, 3, 4]
key = 2
print(lastOccurance(arr, key))
```

```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

# Count occurrences using first and last occurrence
def count(arr, ele):
    fst = firstOccurance(arr, ele)
    
    if fst == -1:   # element not found
        return 0
    
    last = lastOccurance(arr, ele)
    
    return last - fst + 1

# Driver code
arr = [1, 2, 2, 2, 3, 4]
ele = 2

print("First Occurrence:", firstOccurance(arr, ele))
print("Last Occurrence:", lastOccurance(arr, ele))
print("Count:", count(arr, ele))
```
