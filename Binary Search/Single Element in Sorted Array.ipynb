```python
# Time Complexity: O(log n)
# Space Complexity: O(1)

def singleNonDuplicate(nums):
    i = 0                      # start pointer
    n = len(nums)
    j = n - 1                  # end pointer

    # if only one element, that is the answer
    if n == 1:
        return nums[0]

    # binary search
    while i <= j:
        mid = i + (j - i) // 2   # find middle index

        # handle left edge
        # if mid is at index 0, we cannot check mid-1
        if mid == 0:
            # if first element is not equal to next → it's single
            if nums[0] != nums[1]:
                return nums[0]
            else:
                # otherwise move right
                i = mid + 1
                continue

        # handle right edge
        # if mid is at last index, we cannot check mid+1
        if mid == n - 1:
            # if last element is not equal to previous → it's single
            if nums[n - 1] != nums[n - 2]:
                return nums[n - 1]
            else:
                # otherwise move left
                j = mid - 1
                continue

        # check if mid is the single element
        # if mid is different from both neighbors → answer found
        if nums[mid] != nums[mid - 1] and nums[mid] != nums[mid + 1]:
            return nums[mid]

        # -parity-based decision
        # before single element:
        # pairs start at even index → (0,1), (2,3), ...
        # after single element:
        # pattern shifts → pairs start at odd index

        if mid % 2 == 0:
            # even index
            if nums[mid] == nums[mid + 1]:
                # correct pair → single is on right side
                i = mid + 2
            else:
                # pair broken → single is on left side
                j = mid - 1
        else:
            # odd index
            if nums[mid] == nums[mid - 1]:
                # correct pair → single is on right side
                i = mid + 1
            else:
                # pair broken → single is on left side
                j = mid - 1

    return -1  # fallback (should not happen for valid input)

nums = [1,1,2,3,3,4,4,8,8]
print(singleNonDuplicate(nums))
```
