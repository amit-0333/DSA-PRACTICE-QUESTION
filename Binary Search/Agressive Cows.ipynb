```python
# Time Complexity: O(n log(max-min))
#   Binary search runs log(max(arr) - min(arr)) times
#   Each check takes O(n)

# Space Complexity: O(1)

def isPossible(arr, c, mid):
    cows = 1
    lastPosition = arr[0]

    # try placing cows greedily
    for i in range(1, len(arr)):
        if arr[i] - lastPosition >= mid:
            cows += 1
            lastPosition = arr[i]

            # if we placed all cows → valid
            if cows == c:
                return True

    return False


def AggressiveCows(arr, c):
    arr.sort()  

    i = 0
    j = arr[-1] - arr[0]
    ans = -1

    while i <= j:
        mid = i + (j - i) // 2

        if isPossible(arr, c, mid):
            ans = mid
            i = mid + 1     # try larger distance
        else:
            j = mid - 1     # reduce distance

    return ans

arr = [1, 2, 4, 8, 9]
c = 3
print(AggressiveCows(arr,c))
```
