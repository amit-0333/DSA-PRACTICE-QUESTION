```python
# Time Complexity (TC): O(rows * cols * 4^L)
# Space Complexity (SC): O(L)

def exist(board, word):
    rows = len(board)
    cols = len(board[0])

    def dfs(i, j, k):
        # If all characters matched
        if k == len(word):
            return True

        # Out of bounds OR mismatch
        if i < 0 or j < 0 or i >= rows or j >= cols or board[i][j] != word[k]:
            return False

        # Mark visited
        temp = board[i][j]
        board[i][j] = '#'

        # Explore 4 directions
        found = (
            dfs(i+1, j, k+1) or
            dfs(i-1, j, k+1) or
            dfs(i, j+1, k+1) or
            dfs(i, j-1, k+1)
        )

        # Backtrack
        board[i][j] = temp

        return found

    # Try every cell
    for i in range(rows):
        for j in range(cols):
            if dfs(i, j, 0):
                return True

    return False


board = [
    ['A','B','C','E'],
    ['S','F','C','S'],
    ['A','D','E','E']
]

print(exist(board, "ABCCED"))  # True
```
