```python
def search(matrix, x):
    # Time Complexity: O(log(m*n))
    # Space Complexity: O(1)

    m = len(matrix)
    n = len(matrix[0])

    i = 0
    j = m * n - 1

    while i <= j:
        mid = i + (j - i) // 2
        midEle = matrix[mid // n][mid % n]

        if midEle == x:
            return True
        elif midEle < x:
            i = mid + 1
        else:
            j = mid - 1

    return False


# Example
matrix = [
    [1,3,5,7],
    [10,11,16,20],
    [23,30,34,60]
]

print(search(matrix, 3))   # True
print(search(matrix, 13))  # False
```
