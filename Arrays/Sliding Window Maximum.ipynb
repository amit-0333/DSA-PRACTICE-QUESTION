```python
# Time Complexity (TC): O(n) → each element added & removed at most once
# Space Complexity (SC): O(k) → deque stores at most k indices

from collections import deque

def maxSlidingWindow(nums, k):
    dq = deque()   # stores indices
    res = []
    n = len(nums)

    for j in range(n):

        # remove smaller elements from back
        # they are useless because current element is bigger
        while dq and nums[dq[-1]] < nums[j]:
            dq.pop()

        dq.append(j)

        # remove elements out of window (front)
        if dq[0] == j - k:
            dq.popleft()

        # start adding result when first window is formed
        if j >= k - 1:
            # front of deque = max element index
            res.append(nums[dq[0]])

    return res

print(maxSlidingWindow(nums = [1,3,-1,-3,5,3,6,7], k = 3))
```
