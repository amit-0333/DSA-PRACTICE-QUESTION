```python
def majorityElement(arr):
    # Time Complexity: O(n)
    # Space Complexity: O(1)

    cand = None
    count = 0

    # Step 1: Find candidate (O(n))
    for num in arr:
        if count == 0:
            cand = num
        if num == cand:
            count += 1
        else:
            count -= 1

    # Step 2: Verify candidate (O(n))
    freq = 0
    for num in arr:
        if num == cand:
            freq += 1

    # Final check (O(1))
    if freq > len(arr) // 2:
        return cand
    else:
        return -1


nums = [2,2,1,1,1,2,2]
print("majority element is:", majorityElement(nums))
```
