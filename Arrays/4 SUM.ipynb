```python
## brute force
## time complexity - O(n**4)
def fourSum(arr):
    n = len(arr)
    res = set()

    for i in range(n):
        for j in range(i+1, n):
            for k in range(j+1, n):
                for l in range(k+1, n):
                    if arr[i] + arr[j] + arr[k] + arr[l] == 0:
                        temp = tuple(sorted([arr[i], arr[j], arr[k], arr[l]]))
                        res.add(temp)

    return list(res)


nums = [1, 0, -1, 0, -2, 2]
print(fourSum(nums))
```

```python
## optimal solution
## time complexity-O(n³)
def fourSum(arr, target=0):
    arr.sort()
    n = len(arr)
    res = []

    for i in range(n):
        # skip duplicates for i
        if i > 0 and arr[i] == arr[i-1]:
            continue

        for j in range(i+1, n):
            # skip duplicates for j
            if j > i+1 and arr[j] == arr[j-1]:
                continue

            k = j + 1
            l = n - 1

            while k < l:
                total = arr[i] + arr[j] + arr[k] + arr[l]

                if total < target:
                    k += 1
                elif total > target:
                    l -= 1
                else:
                    res.append([arr[i], arr[j], arr[k], arr[l]])
                    k += 1
                    l -= 1

                    # skip duplicates
                    while k < l and arr[k] == arr[k-1]:
                        k += 1
                    while k < l and arr[l] == arr[l+1]:
                        l -= 1

    return res


nums = [1, 0, -1, 0, -2, 2]
print(fourSum(nums))
```
