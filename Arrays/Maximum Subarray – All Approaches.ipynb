```python
## brute force 
## time complexity - O(n³)
def maxSubArray(nums):
    n = len(nums)
    ans = float('-inf')

    for i in range(n):
        for j in range(i, n):
            curr = 0
            for k in range(i, j+1):
                curr += nums[k]
            ans = max(ans, curr)

    return ans

nums = [-2,1,-3,4,-1,2,1,-5,4]
print(maxSubArray(nums))
```

```python
##Better Approach (O(n²))
def maxSubArray(nums):
    n = len(nums)
    ans = float('-inf')

    for i in range(n):
        curr = 0
        for j in range(i, n):
            curr += nums[j]
            ans = max(ans, curr)

    return ans

nums = [-2,1,-3,4,-1,2,1,-5,4]
print(maxSubArray(nums))
```

```python
##Kadane’s Algorithm (O(n))

def maxSubArray(arr):
    curr=0
    mx=float('-inf')
    for i in range(len(arr)):
        curr+=arr[i]
        mx=max(mx,curr)
        if curr<0:
            curr=0
    return mx

nums = [-2,1,-3,4,-1,2,1,-5,4]
print(maxSubArray(nums))
```
