```python
def spiralOrder(matrix):
    # Time Complexity: O(m * n)
    # Space Complexity: O(1) (excluding output list)

    m = len(matrix)
    n = len(matrix[0])

    st_row, end_row = 0, m - 1
    st_col, end_col = 0, n - 1

    res = []

    while st_row <= end_row and st_col <= end_col:

        # top row
        for j in range(st_col, end_col + 1):
            res.append(matrix[st_row][j])
        st_row += 1

        # right column
        for i in range(st_row, end_row + 1):
            res.append(matrix[i][end_col])
        end_col -= 1

        # bottom row
        if st_row <= end_row:
            for j in range(end_col, st_col - 1, -1):
                res.append(matrix[end_row][j])
            end_row -= 1

        # left column
        if st_col <= end_col:
            for i in range(end_row, st_row - 1, -1):
                res.append(matrix[i][st_col])
            st_col += 1

    return res


# Example
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(spiralOrder(matrix))
```
