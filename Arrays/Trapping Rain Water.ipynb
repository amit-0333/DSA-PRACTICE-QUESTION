```python
# Time Complexity (TC): O(n) → multiple linear passes
# Space Complexity (SC): O(n) → using extra arrays (lmax, rmax, water)

def trap(height):
    n = len(height)

    # create arrays to store left max, right max and water
    lmax = [0]*n   # stores max height from left
    rmax = [0]*n   # stores max height from right
    water = [0]*n  # stores water at each index

    # fill left max array
    # lmax[i] = max(height[0] ... height[i])
    lmax[0] = height[0]
    for i in range(1, n):
        lmax[i] = max(lmax[i-1], height[i])

    # fill right max array
    # rmax[i] = max(height[i] ... height[n-1])
    rmax[n-1] = height[n-1]
    for i in range(n-2, -1, -1):
        rmax[i] = max(rmax[i+1], height[i])

    # calculate trapped water
    total = 0
    for i in range(n):
        # water at i = min(left max, right max) - height[i]
        water[i] = min(lmax[i], rmax[i]) - height[i]
        total += water[i]

    return total

height = [0,1,0,2,1,0,1,3,2,1,2,1]
print(trap(height))
```
