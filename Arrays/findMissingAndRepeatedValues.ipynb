```python
def findMissingAndRepeatedValues(grid):
    # Time Complexity: O(n^2)
    # Space Complexity: O(n^2)  (due to set)

    n = len(grid)
    repeated = -1
    missing = -1
    seen = set()

    # Step 1: Find repeated number
    for row in grid:              # O(n)
        for num in row:           # O(n)
            if num in seen:
                repeated = num
            else:
                seen.add(num)

    # Step 2: Find missing number
    for i in range(1, n*n + 1):   # O(n^2)
        if i not in seen:
            missing = i
            break

    return [repeated, missing]


# Example
grid = [
    [1, 3],
    [2, 2]
]

print("Result:", findMissingAndRepeatedValues(grid))
```
