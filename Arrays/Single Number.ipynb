```python
def singleNum(arr):
    # Time Complexity: O(n)
    # Space Complexity: O(n)

    freq = {}

    # Step 1: Count frequency
    for num in arr:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1

    # Step 2: Find element with freq = 1
    for num in freq:
        if freq[num] == 1:
            return num
```

```python
def singleNum(arr):
    # Time Complexity: O(n)
    # Space Complexity: O(1)

    result = 0
    for num in arr:
        result ^= num
    return result
```
