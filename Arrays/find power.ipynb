```python
def findPower(a, n):
    # Time Complexity: O(log n)
    # Space Complexity: O(log n) (recursive stack)

    if n == 0:
        return 1   
    elif n == 1:
        return a
    else:
        mid = n // 2
        res = findPower(a, mid)

    if n % 2 == 0:
        return res * res
    else:
        return res * res * a


a = 2
n = 7
print(findPower(a, n))
```

```python
def findPower(a, n):
    # Time: O(log n), Space: O(1)
    result = 1
    while n > 0:
        if n % 2 == 1:
            result *= a
        a *= a
        n //= 2
    return result

a = 2
n = 7
print(findPower(a, n))
```
