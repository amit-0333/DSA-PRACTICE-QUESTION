```python
##Brute Force Code (O(n³))
def threeSUM(arr):
    n=len(arr)
    res=set()
    for i in range(n):
        for j in range(i+1,n):
            for k in range(j+1,n):
                if arr[i]+arr[j]+arr[k]==0:
                    temp=tuple(sorted((arr[i],arr[j],arr[k])))
                    res.add(temp)
    return res


nums = [-1,0,1,2,-1,-4]
print(threeSUM(nums))
```

```python
## optimal solution
## time complexity-O(n²)

def threeSum(arr):
    n=len(arr)
    arr.sort()   
    res=[]
    for i in range(n):
        if i>0 and arr[i]==arr[i-1]:
            continue
        j=i+1
        k=n-1
        while j<k:
            total=arr[i]+arr[j]+arr[k]
            if total>0:
                k-=1
            elif total<0:
                j+=1
            else:
                res.append((arr[i],arr[j],arr[k]))
                k-=1
                j+=1
                while j<k and arr[j]==arr[j-1]:
                    j+=1
                while j<k and arr[k]==arr[k+1]:
                    k-=1
    return res

nums = [-1,0,1,2,-1,-4]
print(threeSum(nums))
```
