```python
def searchMatrix(matrix, target):
    # Edge case: empty matrix or empty row
    if not matrix or not matrix[0]:
        return False

    # r = number of rows, c = number of columns
    r, c = len(matrix), len(matrix[0])

    # Start from top-right corner
    i, j = 0, c - 1

    # Traverse while indices are within bounds
    while i < r and j >= 0:
        # If element found
        if matrix[i][j] == target:
            return True
        
        # If current element is greater than target
        elif matrix[i][j] > target:
            j -= 1   # move left
        
        # If current element is smaller than target
        else:
            i += 1   # move down

    # Target not found
    return False


matrix = [
 [1,4,7,11,15],
 [2,5,8,12,19],
 [3,6,9,16,22],
 [10,13,14,17,24],
 [18,21,23,26,30]
]

print(searchMatrix(matrix, 6))  
```
