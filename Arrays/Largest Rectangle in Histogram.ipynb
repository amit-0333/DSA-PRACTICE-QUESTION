```python

# Time Complexity (TC): O(n) → each element pushed & popped once
# Space Complexity (SC): O(n) → stack + result array
def nsl(heights):
    pseudo_idx = -1        # when no smaller element on left
    stack = []             # stores (value, index)
    n = len(heights)
    res = []               # stores index of nearest smaller to left

    for i in range(n):

        # if stack empty → no smaller element
        if len(stack) == 0:
            res.append(pseudo_idx)

        # if top is smaller → nearest smaller found
        elif stack[-1][0] < heights[i]:
            res.append(stack[-1][1])

        # pop until smaller element is found
        else:
            while len(stack) > 0 and stack[-1][0] >= heights[i]:
                stack.pop()

            # if no smaller element exists
            if len(stack) == 0:
                res.append(pseudo_idx)
            else:
                res.append(stack[-1][1])

        # push current element
        stack.append((heights[i], i))

    # convert index → distance
    for i in range(n):
        res[i] = i - res[i]

    return res


# Time Complexity (TC): O(n)
# Space Complexity (SC): O(n)
def nsr(heights):
    n = len(heights)
    pseudo_idx = n         # when no smaller element on right
    stack = []
    res = []               # stores index of nearest smaller to right

    for i in range(n-1, -1, -1):

        if len(stack) == 0:
            res.append(pseudo_idx)

        elif stack[-1][0] < heights[i]:
            res.append(stack[-1][1])

        else:
            while len(stack) > 0 and stack[-1][0] >= heights[i]:
                stack.pop()

            if len(stack) == 0:
                res.append(pseudo_idx)
            else:
                res.append(stack[-1][1])

        stack.append((heights[i], i))

    res.reverse()

    # convert index → distance
    for i in range(n):
        res[i] = res[i] - i

    return res


# Time Complexity (TC): O(n)
# Space Complexity (SC): O(n)
def largestRectangleArea(heights):
    n = len(heights)

    # edge case
    if n == 0:
        return 0

    # get left & right distances
    left = nsl(heights)
    right = nsr(heights)

    # compute width for each bar
    width = [0]*n
    for i in range(n):
        width[i] = left[i] + right[i] - 1

    # compute area
    area = [0]*n
    for i in range(n):
        area[i] = heights[i] * width[i]

    # find maximum area
    mx = area[0]
    for i in range(1, n):
        if area[i] > mx:
            mx = area[i]

    return mx

heights = [2,1,5,6,2,3]
print(largestRectangleArea(heights))
```

```python
# Time Complexity (TC): O(n)
# Space Complexity (SC): O(n)

def largestRectangleArea(heights):
    stack = []      # store indices
    max_area = 0
    n = len(heights)

    for i in range(n):

        # when current height is smaller → process stack
        while stack and heights[stack[-1]] > heights[i]:
            h = heights[stack.pop()]   # height of rectangle

            # calculate width
            if not stack:
                w = i
            else:
                w = i - stack[-1] - 1

            max_area = max(max_area, h * w)

        stack.append(i)

    # process remaining bars
    while stack:
        h = heights[stack.pop()]

        if not stack:
            w = n
        else:
            w = n - stack[-1] - 1

        max_area = max(max_area, h * w)

    return max_area

heights = [2,1,5,6,2,3]
print(largestRectangleArea(heights))
```
