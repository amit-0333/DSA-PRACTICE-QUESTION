```python
class Solution:

    # TC: O(N)
    # Each node is visited exactly once

    # SC: O(H)
    # Recursion stack (H = height of tree)

    def toSumTree(self, root: Optional[TreeNode]) -> int:

        if root is None:
            return 0

        # Recursively get sum of left and right subtree
        left = self.toSumTree(root.left)
        right = self.toSumTree(root.right)

        # Store original value
        old_val = root.val

        # Update node to sum of subtree (excluding itself)
        root.val = left + right

        # Return total sum of subtree including original node
        return old_val + root.val
```
