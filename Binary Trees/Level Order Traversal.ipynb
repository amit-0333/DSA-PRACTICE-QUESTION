```python
class Solution:
    
    # TC: O(N)
    # Each node is visited exactly once
    
    # SC: O(H)
    # Recursive call stack space (H = height of tree)
    # In worst case (skewed tree) → O(N), in balanced → O(log N)
    
    def lvl(self, root, res, level):
        
        # Base case: if node is null, stop recursion
        if root is None:
            return 
        
        # If this level is visited first time, create a new list
        if level == len(res):
            res.append([])
        
        # Add current node value to its level
        res[level].append(root.val)
        
        # Recur for left subtree (next level)
        self.lvl(root.left, res, level + 1)
        
        # Recur for right subtree (next level)
        self.lvl(root.right, res, level + 1)

    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        
        # Result list to store level-wise nodes
        res = []
        
        # Start DFS from level 0
        self.lvl(root, res, 0)
        
        return res
```
