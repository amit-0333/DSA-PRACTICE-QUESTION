```python
class Solution:
    
    # TC: O(N)
    # Each node is visited once
    
    # SC: O(H)
    # Recursive stack space (H = height of tree)
    
    def inorder(self, root, res):
        
        # Base case: stop recursion
        if root is None:
            return
        
        # Left subtree
        self.inorder(root.left, res)
        
        # Root
        res.append(root.val)
        
        # Right subtree
        self.inorder(root.right, res)

    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        
        res = []
        
        self.inorder(root, res)
        
        return res
```

```python
class Solution:
    
    # TC: O(N)
    # Each node is pushed and popped exactly once
    
    # SC: O(H)
    # Stack space (H = height of tree)
    # Worst case skewed tree → O(N), balanced → O(log N)
    
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        
        res = []      # To store inorder result
        stack = []    # Stack to simulate recursion
        curr = root   # Start from root
        
        while True:
            
            # Go as left as possible
            if curr is not None:
                stack.append(curr)
                curr = curr.left
            
            # If no left, process node and go right
            elif len(stack) > 0:
                curr = stack.pop()
                
                # Visit node
                res.append(curr.val)
                
                # Move to right subtree
                curr = curr.right
            
            # If stack empty and curr is None → done
            else:
                break
        
        return res
```
