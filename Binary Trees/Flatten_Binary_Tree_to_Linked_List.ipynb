```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:

    # TC: O(N^2) worst case (due to repeated traversal to tail)
    # SC: O(H) recursion stack

    def flatten(self, root: [TreeNode]) -> None:

        if root is None:
            return None

        # Flatten left and right subtree
        self.flatten(root.left)
        self.flatten(root.right)

        # Store right subtree
        temp = root.right

        # Move left subtree to right
        root.right = root.left
        root.left = None

        # Go to the end of new right subtree
        curr = root
        while curr.right:
            curr = curr.right

        # Attach original right subtree
        curr.right = temp

```

```python
# Helper: Print flattened tree (right-skewed list)
def print_flattened(root):
    curr = root
    while curr:
        print(curr.val, end=" -> ")
        curr = curr.right
    print("None")


# Driver Code
if __name__ == "__main__":

    # Construct Binary Tree
    #        1
    #       / \
    #      2   5
    #     / \   \
    #    3   4   6

    root = TreeNode(1)
    root.left = TreeNode(2)
    root.right = TreeNode(5)
    root.left.left = TreeNode(3)
    root.left.right = TreeNode(4)
    root.right.right = TreeNode(6)

    sol = Solution()
    sol.flatten(root)

    print("Flattened Tree:")
    print_flattened(root)
```

```python
class Solution:

    # TC: O(N)
    # Each node is pushed and popped exactly once

    # SC: O(N)
    # Stack can hold up to N nodes in worst case (skewed tree)

    def flatten(self, root: Optional[TreeNode]) -> None:

        if root is None:
            return None

        # Stack for iterative preorder traversal
        s = []
        s.append(root)

        while s:
            curr = s.pop()

            # Push right first so left is processed first
            if curr.right:
                s.append(curr.right)

            if curr.left:
                s.append(curr.left)

            # Link current node to next node in preorder
            if s:
                curr.right = s[-1]

            # Set left to None as required
            curr.left = None
```

```python
# Helper: Print flattened tree (right-skewed list)
def print_flattened(root):
    curr = root
    while curr:
        print(curr.val, end=" -> ")
        curr = curr.right
    print("None")


# Driver Code
if __name__ == "__main__":

    # Construct Binary Tree
    #        1
    #       / \
    #      2   5
    #     / \   \
    #    3   4   6

    root = TreeNode(1)
    root.left = TreeNode(2)
    root.right = TreeNode(5)
    root.left.left = TreeNode(3)
    root.left.right = TreeNode(4)
    root.right.right = TreeNode(6)

    sol = Solution()
    sol.flatten(root)

    print("Flattened Tree:")
    print_flattened(root)
```
