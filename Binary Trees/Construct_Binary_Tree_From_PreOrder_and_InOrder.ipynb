```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# TC: O(N^2)
# Each recursive call uses inorder.index() → O(N)

# SC: O(N)
# Due to recursion + list slicing

def buildTree(self, preorder, inorder):

    # Base case: if lists are empty
    if not preorder or not inorder:
        return None

    # First element of preorder is root
    root = TreeNode(preorder[0])

    # Find root index in inorder
    mid = inorder.index(preorder[0])

    # Build left subtree
    # Left part of inorder → elements before mid
    # Corresponding preorder slice
    root.left = self.buildTree(preorder[1:mid+1], inorder[:mid])

    # Build right subtree
    # Right part of inorder → elements after mid
    root.right = self.buildTree(preorder[mid+1:], inorder[mid+1:])

    return root
```

```python
class Solution:

    # TC: O(N)
    # Each node is processed once

    # SC: O(N)
    # Hashmap + recursion stack

    def buildTree(self, preorder, inorder):

        # Build value -> index map
        self.inorder_map = {val: i for i, val in enumerate(inorder)}

        # Pointer for preorder
        self.pre_idx = 0

        # Call helper
        return self._helper(preorder, 0, len(inorder) - 1)


    def _helper(self, preorder, left, right):

        # Base case
        if left > right:
            return None

        # Pick current root
        root_val = preorder[self.pre_idx]
        self.pre_idx += 1

        root = TreeNode(root_val)

        # Find index in inorder
        mid = self.inorder_map[root_val]

        # Build left subtree
        root.left = self._helper(preorder, left, mid - 1)

        # Build right subtree
        root.right = self._helper(preorder, mid + 1, right)

        return root
```
