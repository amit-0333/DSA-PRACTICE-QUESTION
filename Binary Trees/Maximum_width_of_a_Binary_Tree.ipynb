```python
from collections import deque

class Solution:

    # TC: O(N)
    # Each node is visited exactly once

    # SC: O(N)
    # Queue stores nodes level by level

    def widthOfBinaryTree(self, root: Optional[TreeNode]) -> int:

        if not root:
            return 0

        q = deque()
        width = 0

        # Store (node, index)
        # Index represents position as if tree is a complete binary tree
        q.append((root, 0))

        while q:
            size = len(q)

            # Leftmost index of current level
            st = q[0][1]

            # Rightmost index of current level
            ed = q[-1][1]

            # Width = last index - first index + 1
            # This includes gaps (null nodes) in between
            width = max(width, ed - st + 1)

            for _ in range(size):
                node, idx = q.popleft()

                # Normalize index to prevent overflow in deep trees
                idx -= st

                # Assign indices like complete binary tree
                # Left child → 2*i + 1
                if node.left:
                    q.append((node.left, 2 * idx + 1))

                # Right child → 2*i + 2
                if node.right:
                    q.append((node.right, 2 * idx + 2))

        return width
```
