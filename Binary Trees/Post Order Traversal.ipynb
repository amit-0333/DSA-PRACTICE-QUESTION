```python
class Solution:
    
    # TC: O(N)
    # Each node is visited exactly once
    
    # SC: O(H)
    # Recursive stack space (H = height of tree)
    
    def postOrder(self, root, res):
        
        # Base case
        if root is None:
            return
        
        # Left subtree
        self.postOrder(root.left, res)
        
        # Right subtree
        self.postOrder(root.right, res)
        
        # Root (visit after children)
        res.append(root.val)

    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        
        res = []
        
        self.postOrder(root, res)
        
        return res
```

```python
class Solution:
    
    # TC: O(N)
    # Each node is pushed and popped once from stacks
    
    # SC: O(N)
    # Two stacks used to store nodes/values
    
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        
        # Edge case: empty tree
        if root is None:
            return []
        
        res = []   # Final result
        s1 = []    # First stack
        s2 = []    # Second stack
        
        # Push root to first stack
        s1.append(root)
        
        # Process nodes
        while s1:
            
            curr = s1.pop()
            
            # Store value in second stack
            s2.append(curr.val)
            
            # Push left first, then right (important for correct order)
            if curr.left:
                s1.append(curr.left)
            
            if curr.right:
                s1.append(curr.right)
        
        # Reverse order using second stack
        while s2:
            res.append(s2.pop())
        
        return res
```
