```python
class Solution:

    def isSameTree(self, p, q):

        # Assigning to shorter variable names for convenience
        l, r = p, q

        # Case 1:
        # If both nodes are None → trees match at this branch
        if not l and not r:
            return True

        # Case 2:
        # If one is None and the other is not → mismatch
        if not l or not r:
            return False

        # Case 3:
        # If values differ → trees are not identical
        if l.val != r.val:
            return False

        # Recursive check:
        # Left subtree must match AND right subtree must match
        return self.isSameTree(l.left, r.left) and self.isSameTree(l.right, r.right)

    def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:

        # Case 1:
        # If subRoot is None → always a subtree
        if subRoot is None:
            return True

        # Case 2:
        # If root becomes None but subRoot is not → not possible
        if root is None:
            return False

        # Case 3:
        # If trees match at current node
        if self.isSameTree(root, subRoot):
            return True

        # Otherwise, check in left OR right subtree
        return self.isSubtree(root.left, subRoot) or self.isSubtree(root.right, subRoot)

# TC: O(N * M)
# For each node in root (N), we may compare with subRoot (M)

# SC: O(H)
# Recursion stack
```
