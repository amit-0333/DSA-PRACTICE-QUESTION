```python
from collections import deque

class Solution:

    # TC: O(N)
    # Each node is visited exactly once

    # SC: O(N)
    # Queue + level deque can store up to N nodes

    def zigzagLevelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:

        # Base case: empty tree
        if not root:
            return []

        # Queue for BFS (level order traversal)
        q = deque([root])

        # Final result
        res = []

        # Flag to control direction
        # 0 → left to right
        # 1 → right to left
        flag = 0

        while q:

            # Number of nodes at current level
            size = len(q)

            # Deque to efficiently insert from both ends
            level = deque()

            for _ in range(size):

                # Pop node from queue
                node = q.popleft()

                # Insert based on direction
                if flag == 0:
                    # Left → Right
                    level.append(node.val)
                else:
                    # Right → Left (reverse order)
                    level.appendleft(node.val)   # O(1)

                # Add children for next level
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)

            # Convert deque to list and store result
            res.append(list(level))

            # Toggle direction for next level
            flag = 1 - flag

        return res
```
