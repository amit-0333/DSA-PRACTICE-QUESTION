```python
class Solution:
    
    # TC: O(N)
    # Inorder traversal visits each node once
    
    # SC: O(H)
    # Recursive stack space
    
    def __init__(self):
        self.prev = None
        self.ans = float('inf')
    
    def inorder(self, root):
        
        if root is None:
            return
        
        # Left
        self.inorder(root.left)
        
        # Process current node
        if self.prev is not None:
            self.ans = min(self.ans, root.val - self.prev)
        
        # Update prev
        self.prev = root.val
        
        # Right
        self.inorder(root.right)

        return self.ans
    
    def minDiffInBST(self, root: Optional[TreeNode]) -> int:
        
        self.inorder(root)
        
        return self.ans
```
