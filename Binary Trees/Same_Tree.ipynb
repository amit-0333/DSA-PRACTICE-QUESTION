```python
class Solution:

    # TC: O(N)
    # Visit each node once (N = number of nodes)

    # SC: O(H)
    # Recursion stack (H = height of tree)

    def isSameTree(self, p, q):

        # Assigning to shorter variable names for convenience
        l, r = p, q

        # Case 1:
        # If both nodes are None → trees match at this branch
        if not l and not r:
            return True

        # Case 2:
        # If one is None and the other is not → mismatch
        if not l or not r:
            return False

        # Case 3:
        # If values differ → trees are not identical
        if l.val != r.val:
            return False

        # Recursive check:
        # Left subtree must match AND right subtree must match
        return self.isSameTree(l.left, r.left) and self.isSameTree(l.right, r.right)
```

```python
from collections import deque

class Solution:

    # TC: O(N)
    # Each node is processed once

    # SC: O(N)
    # Queue can hold up to N nodes in worst case

    def isSameTree(self, p, q):

        # Queue will store nodes in pairs (l, r)
        queue = deque()

        # Start with root nodes of both trees
        queue.append(p)
        queue.append(q)

        while queue:

            # Pop two nodes at a time for comparison
            l = queue.popleft()
            r = queue.popleft()

            # Case 1: both are None → valid, continue
            if not l and not r:
                continue

            # Case 2: one is None → mismatch
            if not l or not r:
                return False

            # Case 3: values differ → mismatch
            if l.val != r.val:
                return False

            # Add children in same order for comparison
            # Left with left, Right with right
            queue.append(l.left)
            queue.append(r.left)

            queue.append(l.right)
            queue.append(r.right)

        # If all comparisons passed
        return True
```
