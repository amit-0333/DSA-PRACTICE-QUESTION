## **This avoids recursion by temporarily modifying the tree:**

1.   For each node, connect its inorder predecessor → current node
2.   This helps you come back without a stack
3.   After using it, restore the tree (remove thread)

```python
class Solution:

    # TC: O(N)
    # Each node is visited at most 2 times

    # SC: O(1)
    # No extra space (no recursion / stack), only pointers used

    def inorderTraversal(self, root):

        res = []              # Stores inorder result
        curr = root           # Start from root

        while curr:

            # Case 1: No left subtree → visit node and go right
            if curr.left is None:
                res.append(curr.val)
                curr = curr.right

            else:
                # Find inorder predecessor (rightmost node in left subtree)
                ip = curr.left
                while ip.right and ip.right != curr:
                    ip = ip.right

                # Case 2: First time visiting this predecessor
                if ip.right is None:
                    ip.right = curr      # Create thread
                    curr = curr.left     # Move left

                # Case 3: Thread already exists → remove it
                else:
                    ip.right = None      # Remove thread
                    res.append(curr.val) # Visit node
                    curr = curr.right    # Move right

        return res
```
