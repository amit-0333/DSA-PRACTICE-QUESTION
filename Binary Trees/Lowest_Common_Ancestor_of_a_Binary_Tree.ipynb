```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:

    # TC: O(N)
    # We visit each node at most once in the worst case.

    # SC: O(H)
    # Recursive call stack, where H = height of tree
    # Worst case (skewed tree): O(N)
    # Best/average (balanced tree): O(log N)

    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':

        # Base case: if node is null
        if root is None:
            return None

        # If current node matches either p or q
        # return it upward
        if root == p or root == q:
            return root

        # Recur on left and right subtree
        left = self.lowestCommonAncestor(root.left, p, q)
        right = self.lowestCommonAncestor(root.right, p, q)

        # If both sides return non-null
        # p and q are found in different subtrees
        # so current node is LCA
        if left and right:
            return root

        # Otherwise return the non-null side
        if not left:
            return right

        return left

# ---------------- DRIVER CODE ----------------

# Build tree:
#         3
#       /   \
#      5     1
#     / \   / \
#    6   2 0   8
#       / \
#      7   4

root = TreeNode(3)
root.left = TreeNode(5)
root.right = TreeNode(1)
root.left.left = TreeNode(6)
root.left.right = TreeNode(2)
root.right.left = TreeNode(0)
root.right.right = TreeNode(8)
root.left.right.left = TreeNode(7)
root.left.right.right = TreeNode(4)

# Select nodes
p = root.left              # Node 5
q = root.left.right.right # Node 4

# Call function
obj = Solution()
ans = obj.lowestCommonAncestor(root, p, q)

print("LCA:", ans.val)
```

```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:

    # TC: O(H)
    # H = height of BST

    # SC: O(H) (recursion stack)

    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':

        if root is None:
            return None

        root_val = root.val
        p_val = p.val
        q_val = q.val

        # both nodes in right subtree
        if p_val > root_val and q_val > root_val:
            return self.lowestCommonAncestor(root.right, p, q)

        # both nodes in left subtree
        if p_val < root_val and q_val < root_val:
            return self.lowestCommonAncestor(root.left, p, q)

        # split point → this is LCA
        return root
```
