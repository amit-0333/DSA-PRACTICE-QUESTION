```python
from collections import deque

class Solution:

    # TC: O(N)
    # Each node is visited exactly once

    # SC: O(W)
    # Queue stores at most one level (W = max width of tree)

    def maxLevelSum(self, root):

        # Edge case: empty tree
        if root is None:
            return 0

        q = deque()
        q.append(root)   # start BFS from root

        lvl = 0          # current level number
        ansLvl = 0       # level with maximum sum

        maxSum = float('-inf')  # handle negative values

        # Standard BFS traversal
        while q:
            lvl += 1
            size = len(q)      # number of nodes at current level
            curr_sum = 0       # sum of current level

            # process all nodes of this level
            for _ in range(size):
                node = q.popleft()

                curr_sum += node.val   # add node value

                # push children into queue
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)

            # update max level sum
            if curr_sum > maxSum:
                maxSum = curr_sum
                ansLvl = lvl

        return ansLvl

# -------- DRIVER CODE --------

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


# Build tree:
#         1
#       /   \
#      7     0
#     / \
#    7  -8

root = TreeNode(1)
root.left = TreeNode(7)
root.right = TreeNode(0)
root.left.left = TreeNode(7)
root.left.right = TreeNode(-8)

# Call function
obj = Solution()
print("Max Level Sum at Level:", obj.maxLevelSum(root))
```
