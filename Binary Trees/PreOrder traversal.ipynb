```python
class Solution:
    
    # TC: O(N)
    # Each node is visited once
    
    # SC: O(H)
    # Recursive stack space (H = height of tree)
    
    def preOrder(self, root, res):
        
        # Base case
        if root is None:
            return
        
        # Root
        res.append(root.val)
        
        # Left
        self.preOrder(root.left, res)
        
        # Right
        self.preOrder(root.right, res)
    
    def preorderTraversal(self, root):
        
        res = []
        self.preOrder(root, res)
        
        return res
```

```python
class Solution:
    
    # TC: O(N)
    # Each node is pushed and popped exactly once from the stack
    
    # SC: O(H)
    # Stack space in worst case = height of tree
    # - O(log N) for balanced tree
    # - O(N) for skewed tree
    
    def preorderTraversal(self, root):
        
        # Result list to store traversal
        res = []
        
        # Stack for iterative traversal
        s = []
        
        # Base case
        if root is None:
            return []
        
        # Push root node into stack
        s.append(root)
        
        # Iterate until stack is empty
        while s:
            
            # Pop the top node
            popped = s.pop()
            
            # Process root (Preorder: Root → Left → Right)
            res.append(popped.val)
            
            # Push right child first (so left is processed first)
            if popped.right:
                s.append(popped.right)
            
            # Push left child
            if popped.left:
                s.append(popped.left)
        
        return res
```
