```python
class Solution:

    # TC: O(N)
    # We visit each node exactly once using reverse inorder traversal

    # SC: O(H)
    # Recursive call stack where H = height of tree
    # Worst case (skewed tree): O(N)
    # Best case (balanced tree): O(log N)

    def __init__(self):
        # This will store the cumulative sum of nodes visited so far
        # (i.e., sum of all greater nodes)
        self.running_sum = 0

    def bstToGst(self, root: Optional[TreeNode]) -> Optional[TreeNode]:

        # Base case: if node is None, just return
        if root is None:
            return None

        # Step 1: Traverse right subtree first
        # Reason: Right subtree contains greater values in BST
        self.bstToGst(root.right)

        # Step 2: Process current node
        # Add current node's value to running sum
        self.running_sum += root.val

        # Update node value to include sum of greater nodes
        root.val = self.running_sum

        # Step 3: Traverse left subtree
        # Left subtree has smaller values
        self.bstToGst(root.left)

        # Return modified root
        return root

##Reverse inorder + running sum = Greater Sum Tree
```
