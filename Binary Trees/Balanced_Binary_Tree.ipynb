```python
class Solution:

    # TC: O(N)
    # Each node visited once

    # SC: O(H)
    # Recursion stack (H = height of tree)

    def __init__(self):
        # Global flag to track if tree is balanced
        self.bal = True

    def height(self, root):

        # Base case: empty node → height = 0
        if root is None:
            return 0

        # Recursively get left subtree height
        l = self.height(root.left)

        # Recursively get right subtree height
        r = self.height(root.right)

        # Check balance condition at current node
        # If difference > 1 → tree is not balanced
        if abs(l - r) > 1:
            self.bal = False   # update global flag

        # Return height to parent
        return 1 + max(l, r)

    def isBalanced(self, root):

        # Reset flag (important if object reused)
        self.bal = True

        # Compute height (updates self.bal internally)
        self.height(root)

        # Final result
        return self.bal
```
