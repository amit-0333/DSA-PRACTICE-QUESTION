```python
class Solution:

    # TC: O(N)
    # Each node is visited exactly once

    # SC: O(H)
    # Recursion stack (H = height of tree)

    def maxPathSum(self, root: Optional[TreeNode]) -> int:

        self.max_sum = float('-inf')   # initialize here

        self.helper(root)

        return self.max_sum


    def helper(self, root):

        if root is None:
            return 0

        # Ignore negative contributions
        left = max(self.helper(root.left), 0)
        right = max(self.helper(root.right), 0)

        # Path passing through current node
        curr_path = left + right + root.val

        # Update global max
        self.max_sum = max(self.max_sum, curr_path)

        # Return path usable by parent
        return root.val + max(left, right)
```
