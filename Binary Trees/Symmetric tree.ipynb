```python
class Solution:
    
    # TC: O(N)
    # Each node is checked once
    
    # SC: O(H)
    # Recursive stack space
    
    def isMirror(self, l, r):
        
        # Both null → symmetric
        if not l and not r:
            return True
        
        # One null → not symmetric
        if not l or not r:
            return False
        
        # Values must match
        if l.val != r.val:
            return False
        
        # Cross check 
        return self.isMirror(l.left, r.right) and self.isMirror(l.right, r.left)

    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        
        if root is None:
            return True
        
        return self.isMirror(root.left, root.right)
```

```python
from collections import deque

class Solution:
    
    # TC: O(N)
    # Each node is checked once
    
    # SC: O(N)
    # Queue stores nodes level-wise
    
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        
        if root is None:
            return True
        
        q = deque()
        
        # push  
        q.append(root.left)
        q.append(root.right)
        
        while q:
            
            l = q.popleft()
            r = q.popleft()
            
            if not l and not r:
                continue
            
            if not l or not r:
                return False
            
            if l.val != r.val:
                return False
            
            # Mirror order (this part was already correct)
            q.append(l.left)
            q.append(r.right)
            q.append(l.right)
            q.append(r.left)
        
        return True
```
