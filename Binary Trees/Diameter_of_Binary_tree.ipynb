```python
class Solution:

    def __init__(self):
        # Initialize diameter with very small value
        # (will be reset later, so this is optional)
        self.diameter = float('-inf')

    def height(self, root):

        # Base case:
        # If node is None → height = 0
        if root is None:
            return 0

        # Recursively compute height of left subtree
        left = self.height(root.left)

        # Recursively compute height of right subtree
        right = self.height(root.right)

        # Update diameter:
        # Longest path passing through current node =
        # left subtree height + right subtree height
        # We keep track of maximum across all nodes
        self.diameter = max(self.diameter, (left + right))

        # Return height of current node to parent
        # 1 (edge to parent) + max height of left/right subtree
        return 1 + max(left, right)

    def diameterOfBinaryTree(self, root):

        # Reset diameter (important if same object reused)
        # Also ensures correct answer for empty/small trees
        self.diameter = 0

        # Start DFS traversal to compute height
        # During this process, diameter gets updated
        self.height(root)

        # Final answer: maximum diameter (in terms of edges)
        return self.diameter
```
