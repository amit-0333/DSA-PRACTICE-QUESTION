```python
class Solution:

    def findIP(self, root):
        # APPROACH:
        # Inorder predecessor = rightmost of left subtree

        # TC: O(H)
        # SC: O(1)

        curr = root.left

        while curr.right:
            curr = curr.right

        return curr

    def deleteNode(self, root: Optional[TreeNode], key: int) -> Optional[TreeNode]:
        # APPROACH:
        # 1. Search node
        # 2. Handle cases:
        #    - leaf → return None
        #    - one child → return that child
        #    - two children → replace with IP

        # TC: O(H)
        # SC: O(H)

        if not root:
            return None

        if key < root.val:
            root.left = self.deleteNode(root.left, key)

        elif key > root.val:
            root.right = self.deleteNode(root.right, key)

        else:
            # node found

            # case 0: leaf node
            if not root.left and not root.right:
                return None

            # case 1: only right child
            if not root.left:
                return root.right

            # case 2: only left child
            if not root.right:
                return root.left

            # case 3: two children
            ip = self.findIP(root)
            root.val = ip.val

            root.left = self.deleteNode(root.left, ip.val)

        return root
```

```python
class Solution:

    def findIS(self, root):
        # APPROACH:
        # Inorder successor = leftmost of right subtree

        # TC: O(H)
        # SC: O(1)

        curr = root.right

        while curr.left:
            curr = curr.left

        return curr

    def deleteNode(self, root: Optional[TreeNode], key: int) -> Optional[TreeNode]:
        # APPROACH:
        # 1. Search node
        # 2. Handle cases:
        #    - leaf → return None
        #    - one child → return that child
        #    - two children → replace with inorder successor

        # TC: O(H)
        # SC: O(H)

        if not root:
            return None

        if key < root.val:
            root.left = self.deleteNode(root.left, key)

        elif key > root.val:
            root.right = self.deleteNode(root.right, key)

        else:
            # node found

            # case 0: leaf
            if not root.left and not root.right:
                return None

            # case 1: only right
            if not root.left:
                return root.right

            # case 2: only left
            if not root.right:
                return root.left

            # case 3: two children → use successor
            succ = self.findIS(root)
            root.val = succ.val

            # IMPORTANT: delete from RIGHT subtree
            root.right = self.deleteNode(root.right, succ.val)

        return root
```
