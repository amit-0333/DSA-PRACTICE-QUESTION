```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class BSTIterator:

    def __init__(self, root):
        # APPROACH:
        # Use stack to simulate inorder traversal (Left → Node → Right)
        # Push all left nodes initially

        # TC: O(H) amortized initialization
        # SC: O(H)

        self.s = []
        self.pushleft(root)

    def pushleft(self, root):
        # APPROACH:
        # Push all left nodes into stack

        # TC: O(H)
        # SC: O(H)

        while root:
            self.s.append(root)
            root = root.left

    def next(self) -> int:
        # APPROACH:
        # Pop top node (next smallest)
        # If it has right subtree → push all left of right subtree

        # TC: O(1) amortized
        # SC: O(H)

        node = self.s.pop()

        if node.right:
            self.pushleft(node.right)

        return node.val

    def hasNext(self) -> bool:
        # APPROACH:
        # If stack not empty → more nodes exist

        # TC: O(1)
        # SC: O(1)

        return len(self.s) > 0
```

```python
# APPROACH:
# Build a BST, initialize iterator, and print elements using next()

# TC: O(N)
# SC: O(H)

# -------- DRIVER --------

# Helper TreeNode (if not already defined)
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


# Create BST
#        7
#       / \
#      3   15
#         /  \
#        9    20

root = TreeNode(7)
root.left = TreeNode(3)
root.right = TreeNode(15)
root.right.left = TreeNode(9)
root.right.right = TreeNode(20)

# Initialize iterator
it = BSTIterator(root)

# Iterate
while it.hasNext():
    print(it.next(), end=" ")

# Expected Output:
# 3 7 9 15 20
```
