```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:

    # APPROACH:
    # 1. We are given a sorted array → we need to convert it into a height-balanced BST.
    #
    # 2. Key idea:
    #    - Middle element of array should be the root (to keep tree balanced).
    #    - Elements on left of mid → form left subtree
    #    - Elements on right of mid → form right subtree
    #
    # 3. Use recursion:
    #    - Pick mid = (l + r) // 2
    #    - Create root node from arr[mid]
    #    - Recursively build:
    #         left subtree → (l, mid-1)
    #         right subtree → (mid+1, r)
    #
    # 4. Base case:
    #    - If l > r → no elements → return None


    # TC: O(N)
    # Each element is used exactly once to create a node

    # SC: O(H)
    # Recursive stack space → O(log N) for balanced tree

    def srt(self, arr, l, r):

        # Base case: no elements left
        if l > r:
            return None

        # Step 1: Choose middle element as root
        mid = l + (r - l) // 2

        # Create root node
        root = TreeNode(arr[mid])

        # Step 2: Recursively build left subtree
        root.left = self.srt(arr, l, mid - 1)

        # Step 3: Recursively build right subtree
        root.right = self.srt(arr, mid + 1, r)

        return root


    def sortedArrayToBST(self, nums):

        # Call helper with full array range
        return self.srt(nums, 0, len(nums) - 1)

# DRIVER CODE

# Helper function to print inorder traversal (to verify BST)
def inorder(root):
    if not root:
        return
    inorder(root.left)
    print(root.val, end=" ")
    inorder(root.right)


# Input
nums = [-10, -3, 0, 5, 9]

# Create Solution object
obj = Solution()

# Build BST
root = obj.sortedArrayToBST(nums)

# Output (should be sorted if BST is correct)
print("Inorder Traversal of BST:")
inorder(root)   # Expected: -10 -3 0 5 9
```
