```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    # APPROACH:
    # 1. Use the property of BST:
    #    - Inorder traversal (Left → Root → Right) gives nodes in sorted (ascending) order.
    #
    # 2. Perform iterative inorder traversal using a stack:
    #    - Go to the leftmost node while pushing nodes into the stack.
    #    - Once no left child exists, pop from stack (this is the next smallest element).
    #    - Increment a counter each time we visit a node.
    #    - When counter == k, return that node's value.
    #
    # 3. After visiting a node, move to its right subtree and repeat the process.
    #
    # 4. We DO NOT store the full traversal → we stop early when kth element is found.


    # TC: O(H + K)
    # H = height of tree (time to reach leftmost node)
    # K = number of nodes processed until kth smallest is found

    # SC: O(H)
    # Stack stores at most H nodes (height of tree)

    def kthSmallest(self, root, k: int) -> int:

        # Stack to simulate recursive inorder traversal
        stack = []

        # Start from root
        curr = root

        # Count of visited nodes
        count = 0

        # Loop until we find kth smallest
        while True:

            # Step 1: Traverse left subtree
            # Keep pushing nodes until reaching the leftmost node
            if curr:
                stack.append(curr)      # Push current node to stack
                curr = curr.left        # Move to left child

            # Step 2: Process node when no more left children
            elif stack:
                curr = stack.pop()      # Pop the top node (next smallest)
                count += 1              # Increment visited node count

                # Step 3: Check if this is the kth smallest element
                if count == k:
                    return curr.val     # Return the answer

                # Step 4: Move to right subtree
                curr = curr.right

            # Step 5: If stack empty and curr is None → traversal complete
            else:
                break

# DRIVER CODE

# Build BST manually:
#        5
#       / \
#      3   6
#     / \
#    2   4
#   /
#  1

root = TreeNode(5)
root.left = TreeNode(3)
root.right = TreeNode(6)
root.left.left = TreeNode(2)
root.left.right = TreeNode(4)
root.left.left.left = TreeNode(1)

k = 3

# Create Solution object
obj = Solution()

# Call function
print(obj.kthSmallest(root, k))   # Expected Output: 3
```
