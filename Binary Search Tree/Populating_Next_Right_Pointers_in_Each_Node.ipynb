```python
class Node:
    def __init__(self, val: int = 0, left: 'Node' = None, right: 'Node' = None, next: 'Node' = None):
        self.val = val
        self.left = left
        self.right = right
        self.next = next

class Solution:
    def connect(self, root: Node):
    # APPROACH:
    # For perfect binary tree:
    # 1. Connect left -> right
    # 2. Connect right -> next's left

    # TC: O(N)
    # SC: O(H)

        if not root:
            return None

        if root.left:
            # connect left -> right
            root.left.next = root.right

            # connect right -> next subtree
            if root.next:
                root.right.next = root.next.left

        self.connect(root.left)
        self.connect(root.right)

        return root
```

```python
# APPROACH:
# Build a sample binary tree, call connect(), and print level order using next pointers

# TC: O(N)
# SC: O(N)

# Helper to print using next pointers
def print_levels(root):
    level_start = root
    while level_start:
        curr = level_start
        next_level_start = None

        while curr:
            print(curr.val, end=" -> ")
            if not next_level_start:
                if curr.left:
                    next_level_start = curr.left
                elif curr.right:
                    next_level_start = curr.right
            curr = curr.next

        print("None")
        level_start = next_level_start


# -------- DRIVER --------

# Create tree
#        1
#       / \
#      2   3
#     / \   \
#    4   5   7

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.right = Node(7)

# Call solution
sol = Solution()
sol.connect(root)

# Print result
print_levels(root)
```

```python
from collections import deque

class Solution:
    def connect(self, root):
        # APPROACH:
        # Use level order traversal (BFS) with a queue.
        # For each level:
        #   - Traverse all nodes in that level
        #   - Connect previous node's next to current node
        #   - Last node's next = None
        #
        # NOTE:
        # The earlier condition:
        #   if not root.left or not root.right: return root
        # was incorrect because it stops processing prematurely.
        # We should process even if tree is not perfect.

        # TC: O(N)
        # Each node is visited exactly once

        # SC: O(N)
        # Queue stores nodes level by level (worst case full level)

        if not root:
            return None

        q = deque([root])

        while q:
            size = len(q)
            prev = None

            for _ in range(size):
                curr = q.popleft()

                if prev:
                    prev.next = curr
                prev = curr

                if curr.left:
                    q.append(curr.left)
                if curr.right:
                    q.append(curr.right)

            # last node in level points to None
            if prev:
                prev.next = None

        return root
```
