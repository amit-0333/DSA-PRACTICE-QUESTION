```python
class Solution:

    def pushLeft(self, root, stack):
        # APPROACH:
        # Push all left nodes into stack

        # TC: O(H)
        # SC: O(H)

        while root:
            stack.append(root)
            root = root.left

    def mergeBST(self, root1, root2):
        # APPROACH:
        # Use two stacks to simulate inorder traversal
        # Compare top elements and build sorted result

        # TC: O(N + M)
        # SC: O(H1 + H2)

        s1, s2 = [], []
        self.pushLeft(root1, s1)
        self.pushLeft(root2, s2)

        res = []

        while s1 or s2:

            # choose smaller top
            if not s2 or (s1 and s1[-1].val <= s2[-1].val):
                node = s1.pop()
                res.append(node.val)
                self.pushLeft(node.right, s1)
            else:
                node = s2.pop()
                res.append(node.val)
                self.pushLeft(node.right, s2)

        return res
```

```python
# APPROACH:
# Build two BSTs and merge them using two-stack inorder method

# TC: O(N + M)
# SC: O(H1 + H2)

# Helper TreeNode (if not already defined)
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


# -------- DRIVER --------

# BST 1
#     3
#    / \
#   1   5
root1 = TreeNode(3)
root1.left = TreeNode(1)
root1.right = TreeNode(5)

# BST 2
#     4
#    / \
#   2   6
root2 = TreeNode(4)
root2.left = TreeNode(2)
root2.right = TreeNode(6)

# Call solution
sol = Solution()
result = sol.mergeBST(root1, root2)

# Print merged sorted list
print(result)   # Expected: [1, 2, 3, 4, 5, 6]
```

```python
class Solution:

    def inorder(self, root, res):
        # APPROACH:
        # Standard inorder traversal to get sorted values

        # TC: O(N)
        # SC: O(H)

        if not root:
            return

        self.inorder(root.left, res)
        res.append(root.val)
        self.inorder(root.right, res)

    def merge(self, a, b):
        # APPROACH:
        # Merge two sorted arrays

        # TC: O(N + M)
        # SC: O(N + M)

        i = j = 0
        res = []

        while i < len(a) and j < len(b):
            if a[i] < b[j]:
                res.append(a[i])
                i += 1
            else:
                res.append(b[j])
                j += 1

        # remaining elements
        while i < len(a):
            res.append(a[i])
            i += 1

        while j < len(b):
            res.append(b[j])
            j += 1

        return res

    def buildBST(self, arr, l, h):
        # APPROACH:
        # Build balanced BST using sorted array

        # TC: O(N)
        # SC: O(H)

        if l > h:
            return None

        mid = (l + h) // 2
        root = TreeNode(arr[mid])

        root.left = self.buildBST(arr, l, mid - 1)
        root.right = self.buildBST(arr, mid + 1, h)

        return root

    def mergeTwoBSTs(self, root1, root2):
        # APPROACH:
        # 1. Get inorder of both BSTs
        # 2. Merge sorted arrays
        # 3. Build balanced BST

        # TC: O(N + M)
        # SC: O(N + M)

        arr1 = []
        arr2 = []

        self.inorder(root1, arr1)
        self.inorder(root2, arr2)

        merged = self.merge(arr1, arr2)

        return self.buildBST(merged, 0, len(merged) - 1)
```
