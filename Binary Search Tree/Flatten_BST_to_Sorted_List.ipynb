```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def flattenBST(root):
    # APPROACH:
    # Perform iterative inorder traversal using a stack.
    # Since inorder of BST is sorted, we visit nodes in sorted order.
    # While visiting:
    #   - Link previous node's right to current node
    #   - Set current node's left to None (flatten)
    #   - Move prev pointer forward
    # First visited node becomes head of the flattened list

    # TC: O(N)
    # Each node is visited exactly once

    # SC: O(H)
    # Stack space where H = height of tree

    head = None   # will store first (smallest) node
    prev = None   # tracks previous node in inorder
    s = []        # stack for traversal
    curr = root   # start from root

    # standard iterative inorder traversal
    while s or curr:

        # go to leftmost node
        while curr:
            s.append(curr)
            curr = curr.left

        # get next node in inorder
        curr = s.pop()

        # if first node, mark as head
        if not head:
            head = curr

        # link previous node to current
        if prev:
            prev.right = curr

        # remove left pointer (flatten)
        curr.left = None

        # move prev forward
        prev = curr

        # move to right subtree
        curr = curr.right

    return head
```

```python
# APPROACH:
# Build a BST, flatten it, and print the right-skewed list

# TC: O(N)
# SC: O(H)


# Helper to print flattened list
def print_list(head):
    while head:
        print(head.val, end=" -> ")
        head = head.right
    print("None")


# -------- DRIVER --------

# Create BST
#        5
#       / \
#      3   7
#     / \   \
#    2   4   8

root = TreeNode(5)
root.left = TreeNode(3)
root.right = TreeNode(7)
root.left.left = TreeNode(2)
root.left.right = TreeNode(4)
root.right.right = TreeNode(8)

# Flatten
head = flattenBST(root)

# Print result
print_list(head)

# Expected:
# 2 -> 3 -> 4 -> 5 -> 7 -> 8 -> None
```

```python


class Solution:
    def flattenBST(self, root):
        # APPROACH:
        # Perform inorder traversal (sorted order)
        # Maintain a prev pointer
        # For each node:
        #   prev.right = curr
        #   curr.left = None
        # Move prev forward

        # TC: O(N)
        # SC: O(H)

        self.prev = None
        self.head = None

        def inorder(node):
            if not node:
                return

            inorder(node.left)

            # process current node
            if self.prev is None:
                self.head = node   # first node (smallest)
            else:
                self.prev.right = node

            node.left = None   # remove left link
            self.prev = node

            inorder(node.right)

        inorder(root)
        return self.head
```
