```python
class Solution:

    # APPROACH:
    # 1. Use the property of BST:
    #    - Inorder traversal (Left → Root → Right) should give a strictly increasing sequence.
    #
    # 2. Perform iterative inorder traversal using a stack:
    #    - Go to the leftmost node while pushing nodes into the stack.
    #    - Pop node → this is the next smallest element.
    #    - Compare with previous value:
    #         if current <= previous → NOT a BST → return False
    #
    # 3. Maintain a variable 'prev' to track the last visited node value.
    #
    # 4. If traversal completes without violation → return True
    #
    # 5. FIXES from your code:
    #    - Do NOT do curr = curr.left after popping (wrong traversal)
    #    - Correct flow:
    #         pop → process → move to right subtree


    # TC: O(N)
    # Each node is visited exactly once

    # SC: O(H)
    # Stack space (height of tree)

    def isValidBST(self, root) -> bool:

        # Stack for inorder traversal
        stack = []

        # Start from root
        curr = root

        # Previous value (initialize to -infinity)
        prev = float('-inf')

        while True:

            # Step 1: Go left as much as possible
            if curr:
                stack.append(curr)
                curr = curr.left

            # Step 2: Process node
            elif stack:
                curr = stack.pop()

                # Check BST condition (strictly increasing)
                if curr.val <= prev:
                    return False

                # Update previous value
                prev = curr.val

                # Move to right subtree
                curr = curr.right

            # Step 3: Traversal finished
            else:
                break

        return True
```

```python
# DRIVER CODE

# Definition for TreeNode
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


# ----------- TEST CASE 1 (Valid BST) -----------
#        5
#       / \
#      3   7
#     / \ / \
#    2  4 6  8

root1 = TreeNode(5)
root1.left = TreeNode(3)
root1.right = TreeNode(7)

root1.left.left = TreeNode(2)
root1.left.right = TreeNode(4)

root1.right.left = TreeNode(6)
root1.right.right = TreeNode(8)


# ----------- TEST CASE 2 (Invalid BST) -----------
#        5
#       / \
#      3   7
#         /
#        4    (should be >5 but is 4)

root2 = TreeNode(5)
root2.left = TreeNode(3)
root2.right = TreeNode(7)
root2.right.left = TreeNode(4)   # violates BST rule


# Create Solution object
obj = Solution()

# Run tests
print(obj.isValidBST(root1))  # Expected: True
print(obj.isValidBST(root2))  # Expected: False
```
