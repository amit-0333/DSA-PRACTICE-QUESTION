```python
from collections import deque

class Codec:

    def serialize(self, root):
        # APPROACH:
        # Use level order traversal (BFS)
        # Store values and 'NULL' for missing nodes
        # Join with commas to form string

        # TC: O(N)
        # Each node is processed once

        # SC: O(N)
        # Queue + result storage

        if not root:
            return ''

        q = deque([root])
        res = []

        while q:
            node = q.popleft()

            if node:
                res.append(str(node.val))
                q.append(node.left)
                q.append(node.right)
            else:
                res.append('NULL')

        return ','.join(res)


    def deserialize(self, data):
        # APPROACH:
        # Reconstruct tree using BFS
        # Use index 'i' to read values from serialized list
        # IMPORTANT:
        #   - val[i] gives next node value
        #   - DO NOT use node.val while creating children

        # TC: O(N)
        # Each value is processed once

        # SC: O(N)
        # Queue storage

        if not data:
            return None

        val = data.split(',')

        root = TreeNode(int(val[0]))   # root from first value
        q = deque([root])
        i = 1

        while q:
            node = q.popleft()

            # left child
            if i < len(val) and val[i] != 'NULL':
                node.left = TreeNode(int(val[i]))
                q.append(node.left)
            i += 1

            # right child
            if i < len(val) and val[i] != 'NULL':
                node.right = TreeNode(int(val[i]))
                q.append(node.right)
            i += 1

        return root
```
