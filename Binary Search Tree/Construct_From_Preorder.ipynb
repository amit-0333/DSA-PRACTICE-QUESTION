```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def build(self, preorder, l, h):
        # APPROACH:
        # Use preorder traversal with bounds (range method).
        # Maintain an index (self.i) to track current element.
        # For each value:
        #   - If it lies within (l, h), create a node
        #   - Recursively build left subtree with (l, val)
        #   - Recursively build right subtree with (val, h)
        #   - If out of bounds, do not consume element (important)

        # TC: O(N)
        # Each element is processed exactly once

        # SC: O(H)
        # Recursive stack space (H = height of tree)

        if self.i == len(preorder):
            return None

        val = preorder[self.i]

        # Use strict bounds to maintain valid BST (no duplicates allowed)
        if val <= l or val >= h:
            return None

        root = TreeNode(val)
        self.i += 1

        root.left = self.build(preorder, l, val)
        root.right = self.build(preorder, val, h)

        return root

    def bstFromPreorder(self, preorder):
        # APPROACH:
        # Initialize index and call recursive builder with full range

        # TC: O(N)
        # SC: O(H)

        self.i = 0
        return self.build(preorder, float('-inf'), float('inf'))
```

```python
# APPROACH:
# Build BST from preorder and verify using inorder traversal (should be sorted)

# TC: O(N)
# SC: O(H)

# Helper: inorder traversal
def inorder(root):
    if root is None:
        return
    inorder(root.left)
    print(root.val, end=" ")
    inorder(root.right)


# -------- DRIVER --------

preorder = [8, 5, 1, 7, 10, 12]

sol = Solution()
root = sol.bstFromPreorder(preorder)

print("Inorder of constructed BST:")
inorder(root)   # Expected: 1 5 7 8 10 12
print()
```
