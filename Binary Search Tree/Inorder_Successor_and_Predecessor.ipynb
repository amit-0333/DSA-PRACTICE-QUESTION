```python
def inorderSuccessor(root, p):
    # APPROACH:
    # Case 1: if right subtree exists → find min in right
    # Case 2: otherwise track ancestor

    # TC: O(H)
    # SC: O(1)

    # Case 1
    if p.right:
        curr = p.right
        while curr.left:
            curr = curr.left
        return curr

    # Case 2
    succ = None
    while root:
        if p.val < root.val:
            succ = root
            root = root.left
        else:
            root = root.right

    return succ
```

```python
def inorderPredecessor(root, p):
    # APPROACH:
    # Case 1: if left exists → max(left)
    # Case 2: else → track ancestor

    # TC: O(H)
    # SC: O(1)

    # Case 1
    if p.left:
        curr = p.left
        while curr.right:
            curr = curr.right
        return curr

    # Case 2
    pred = None
    while root:
        if p.val > root.val:
            pred = root
            root = root.right
        else:
            root = root.left

    return pred
```
