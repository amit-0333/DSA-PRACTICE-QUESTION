```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def canMerge(self, trees):
        # APPROACH:
        # 1. Map root values to tree nodes
        # 2. Count occurrences of all values
        # 3. Find unique root (appears once)
        # 4. DFS merge trees at leaf nodes
        # 5. Validate BST using bounds

        # TC: O(N)
        # SC: O(N)

        from collections import Counter

        root_map = {t.val: t for t in trees}
        count = Counter()

        # count all node values
        for t in trees:
            count[t.val] += 1
            if t.left:
                count[t.left.val] += 1
            if t.right:
                count[t.right.val] += 1

        # find unique root
        root = None
        for t in trees:
            if count[t.val] == 1:
                root = t
                break

        if not root:
            return None

        # remove chosen root from map
        root_map.pop(root.val)

        def dfs(node, low, high):
            # APPROACH:
            # Merge and validate BST using bounds

            # TC: O(N)
            # SC: O(H)

            if not node:
                return True

            if node.val <= low or node.val >= high:
                return False

            # if leaf and can merge
            if not node.left and not node.right and node.val in root_map:
                merge_node = root_map.pop(node.val)
                node.left = merge_node.left
                node.right = merge_node.right

            return dfs(node.left, low, node.val) and dfs(node.right, node.val, high)

        # validate + merge
        if not dfs(root, float('-inf'), float('inf')):
            return None

        # ensure all trees used
        if root_map:
            return None

        return root
```

```python
# APPROACH:
# Create multiple small BSTs, attempt to merge them using canMerge(),
# and print inorder of final tree (should be sorted if valid)

# TC: O(N)
# SC: O(H)

# Helper: inorder traversal
def inorder(root):
    if not root:
        return
    inorder(root.left)
    print(root.val, end=" ")
    inorder(root.right)


# -------- DRIVER --------

# Example:
# Tree1:   2
#         / \
#        1   3
#
# Tree2:   3
#             \
#              5
#
# Tree3:   5
#         /
#        4

t1 = TreeNode(2, TreeNode(1), TreeNode(3))
t2 = TreeNode(3, None, TreeNode(5))
t3 = TreeNode(5, TreeNode(4), None)

trees = [t1, t2, t3]

sol = Solution()
root = sol.canMerge(trees)

if root:
    print("Merged BST Inorder:")
    inorder(root)   # Expected: 1 2 3 4 5
    print()
else:
    print("Cannot merge into valid BST")
```
