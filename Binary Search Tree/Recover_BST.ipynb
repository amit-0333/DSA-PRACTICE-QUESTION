```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    def Inorder(self, root, res):
        # APPROACH:
        # Perform inorder traversal and store values

        # TC: O(N)
        # SC: O(H)

        if root is None:
            return res   # always return list

        self.Inorder(root.left, res)
        res.append(root.val)
        self.Inorder(root.right, res)
        return res

    def fix(self, root, a, b):
        # APPROACH:
        # Traverse tree and swap incorrect node values

        # TC: O(N)
        # SC: O(H)

        if root is None:
            return

        if root.val == a:
            root.val = b
        elif root.val == b:
            root.val = a

        self.fix(root.left, a, b)
        self.fix(root.right, a, b)

    def recoverTree(self, root):
        # APPROACH:
        # 1. Get inorder traversal (should be sorted)
        # 2. Find two misplaced values
        # 3. Swap them in tree

        # TC: O(N)
        # SC: O(N)

        lst = self.Inorder(root, [])
        n = len(lst)

        # find first wrong element
        a = lst[0]
        for i in range(1, n):
            if lst[i] < lst[i - 1]:
                a = lst[i - 1]
                break

        # find second wrong element
        b = lst[n - 1]
        for i in range(n - 2, -1, -1):
            if lst[i] > lst[i + 1]:
                b = lst[i + 1]
                break

        # swap in actual tree
        self.fix(root, a, b)


# APPROACH:
# Build a sample tree, print inorder before and after calling recoverTree

# TC: O(N)
# SC: O(H)

# Helper: inorder traversal
def inorder(root):
    if root is None:
        return
    inorder(root.left)
    print(root.val, end=" ")
    inorder(root.right)


# -------- DRIVER --------

# Create tree with swapped nodes
root = TreeNode(3)
root.left = TreeNode(1)
root.right = TreeNode(2)

print("Before fix:")
inorder(root)
print()

# Call solution
sol = Solution()
sol.recoverTree(root)

print("After fix:")
inorder(root)
print()
```

```python
class Solution:
    def recoverTree(self, root):
    # APPROACH:
    # Iterative inorder traversal using stack
    # Detect two nodes where BST property breaks
    # Swap their values

    # TC: O(N)
    # SC: O(H)

        if root is None:
            return

        s = []
        curr = root
        prev = None
        fst = None
        snd = None

        while s or curr:

            # go left
            while curr:
                s.append(curr)
                curr = curr.left

            curr = s.pop()

            # detect violation
            if prev and prev.val > curr.val:
                if fst is None:
                    fst = prev
                snd = curr

            prev = curr
            curr = curr.right

        # swap values
        if fst and snd:
            fst.val, snd.val = snd.val, fst.val

# APPROACH:
# Build a sample tree, print inorder before and after calling recoverTree

# TC: O(N)
# SC: O(H)

# Helper: inorder traversal
def inorder(root):
    if root is None:
        return
    inorder(root.left)
    print(root.val, end=" ")
    inorder(root.right)


# -------- DRIVER --------

# Create tree with swapped nodes
root = TreeNode(3)
root.left = TreeNode(1)
root.right = TreeNode(2)

print("Before fix:")
inorder(root)
print()

# Call solution
sol = Solution()
sol.recoverTree(root)

print("After fix:")
inorder(root)
print()
```
