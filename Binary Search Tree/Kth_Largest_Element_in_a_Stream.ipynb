```python
import heapq

class KthLargest:

    # TC:
    # __init__: O(N log N) due to heapify + popping extra elements
    # add(): O(log K) for push + possible pop

    # SC: O(K)
    # Heap stores at most k elements

    def __init__(self, k, nums):

        # Store k
        self.k = k

        # Initialize heap with given numbers
        self.heap = nums

        # Convert list into min heap
        heapq.heapify(self.heap)

        # Keep only k largest elements
        while len(self.heap) > self.k:
            heapq.heappop(self.heap)

    def add(self, val: int) -> int:

        # Push new value into heap
        heapq.heappush(self.heap, val)

        # If heap exceeds size k, remove smallest
        if len(self.heap) > self.k:
            heapq.heappop(self.heap)

        # Root of heap = kth largest element
        return self.heap[0]

# Driver Code

# Example:
# k = 3, nums = [4, 5, 8, 2]
# add operations → [3, 5, 10, 9, 4]

k = 3
nums = [4, 5, 8, 2]

# Initialize object
obj = KthLargest(k, nums)

# Perform operations
print(obj.add(3))   # Expected: 4
print(obj.add(5))   # Expected: 5
print(obj.add(10))  # Expected: 5
print(obj.add(9))   # Expected: 8
print(obj.add(4))   # Expected: 8
```

```python
import heapq

class KthLargest:

    # TC:
    # __init__: O(N log K)
    # add(): O(log K)

    # SC: O(K)
    # Heap stores only k elements

    def __init__(self, k: int, nums):

        self.k = k
        self.heap = []

        # Build heap with at most k elements
        for num in nums:
            heapq.heappush(self.heap, num)

            # Maintain size k
            if len(self.heap) > self.k:
                heapq.heappop(self.heap)


    def add(self, val: int) -> int:

        # Add new value
        heapq.heappush(self.heap, val)

        # Remove smallest if size exceeds k
        if len(self.heap) > self.k:
            heapq.heappop(self.heap)

        # kth largest = smallest in heap
        return self.heap[0]

# Driver Code

# Example:
# k = 3, nums = [4, 5, 8, 2]
# add operations → [3, 5, 10, 9, 4]

k = 3
nums = [4, 5, 8, 2]

# Initialize object
obj = KthLargest(k, nums)

# Perform operations
print(obj.add(3))   # Expected: 4
print(obj.add(5))   # Expected: 5
print(obj.add(10))  # Expected: 5
print(obj.add(9))   # Expected: 8
print(obj.add(4))   # Expected: 8
```

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


class KthLargest:

    # TC:
    # add(): O(H + K)
    # H = height of tree, K = traversal steps

    # SC: O(H)
    # Stack space for traversal

    def __init__(self, k: int, nums):

        self.k = k
        self.root = None

        # Build BST from initial nums
        for num in nums:
            self.root = self.insert(self.root, num)


    # Helper: Insert into BST
    def insert(self, root, val):

        if root is None:
            return TreeNode(val)

        if val < root.val:
            root.left = self.insert(root.left, val)
        else:
            root.right = self.insert(root.right, val)

        return root


    def add(self, val: int) -> int:

        # Step 1: Insert new value into BST
        self.root = self.insert(self.root, val)

        # Step 2: Reverse inorder (Right → Root → Left)
        stack = []
        curr = self.root
        count = 0

        while True:

            # Go to rightmost node
            if curr:
                stack.append(curr)
                curr = curr.right

            # Process node
            elif stack:
                curr = stack.pop()
                count += 1

                # kth largest found
                if count == self.k:
                    return curr.val

                # Move to left subtree
                curr = curr.left

            else:
                break

# Driver Code

# Input:
k = 3
nums = [4, 5, 8, 2]

# Initialize object
obj = KthLargest(k, nums)

# Perform add operations
print(obj.add(3))    # Expected: 4
print(obj.add(5))    # Expected: 5
print(obj.add(10))   # Expected: 5
print(obj.add(9))    # Expected: 8
print(obj.add(4))    # Expected: 8
```

```python
class Solution:

    # TC: O(H + K)
    # H = height of tree (going right), then K nodes traversal

    # SC: O(H)
    # Stack for traversal

    def kthLargest(self, root, k: int) -> int:

        stack = []
        curr = root
        count = 0

        while True:

            # Step 1: Go as right as possible
            if curr:
                stack.append(curr)
                curr = curr.right

            # Step 2: Process node
            elif stack:
                curr = stack.pop()
                count += 1

                # Step 3: Check if kth largest
                if count == k:
                    return curr.val

                # Step 4: Move to left subtree
                curr = curr.left

            else:
                break

# Driver Code

# Input:
k = 3
nums = [4, 5, 8, 2]

# Initialize object
obj = KthLargest(k, nums)

# Perform add operations
print(obj.add(3))    # Expected: 4
print(obj.add(5))    # Expected: 5
print(obj.add(10))   # Expected: 5
print(obj.add(9))    # Expected: 8
print(obj.add(4))    # Expected: 8
```
